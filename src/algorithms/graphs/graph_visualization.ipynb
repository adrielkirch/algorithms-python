# Introduction to Graphs in Computer Science

## What is a Graph?

In computer science, a **graph** is a data structure that consists of a set of **nodes** (vertices) and a set of **edges** that connect pairs of nodes. This structure is used to represent relationships or connections between various entities.

## Nodes and Edges

- **Nodes (Vertices):** Represent entities such as people, cities, web pages, or any other discrete items.
- **Edges:** Represent the relationships or connections between nodes. Depending on the type of graph, edges can be directed or undirected.

## Types of Graphs

### Directed and Undirected Graphs

- **Undirected Graph:** Edges have no direction. If there is an edge between nodes A and B, you can traverse from A to B or from B to A without any restriction.



- **Directed Graph (Digraph):** Edges have a direction. If there is a directed edge from node A to node B, it means you can travel from A to B but not necessarily from B to A unless there's another directed edge in the opposite direction.
  


### Weighted and Unweighted Graphs

- **Unweighted Graph:** All edges are considered to have the same weight or cost.



- **Weighted Graph:** Each edge is assigned a weight or cost, which can represent distances, time, costs, or any other relevant metric associated with the relationship between nodes.



### Connectivity and Paths

- **Connectivity:** Refers to how nodes are connected within a graph. A graph can be connected (every pair of nodes is connected by at least one path) or disconnected (some pairs of nodes are not connected by any path).

- **Paths:** A path in a graph is a sequence of edges that allows you to travel from one node to another. Paths can be simple (no repeated nodes) or cyclic (forming a loop).

### Arrows and Loops

- **Arrows:** In directed graphs, arrows (or directed edges) indicate the direction of the relationship between nodes.

- **Loops:** A loop occurs when an edge connects a node to itself. This is more common in certain types of graphs and can represent self-referential relationships.

## Applications of Graphs

Graphs are widely used in various applications, including:

- **Social Networks:** Representing friendships or relationships between individuals.
- **Routing Algorithms:** Finding the shortest path between nodes in networks (e.g., GPS navigation, maze).
- **Neural networking:** Convolutional Neural Networks (CNNs) can be understood and represented as interconnected graphs.
- **Dependency Resolution:** Managing dependencies between software components.
- **Recommendation Systems:** Suggesting products or content based on user preferences and relationships.

Understanding graphs and their properties is fundamental in computer science, enabling efficient representation and manipulation of complex relationships and structures.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()
# Add edges to the graph
G.add_edges_from([("B", "A"),("A", "B"),("D", "B"), ("C", "D"), ("D", "A"), ("A", "C"),("B", "B"), ])

# Draw the graph with arrows
nx.draw(G, with_labels=True, arrows=True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()
# Add edges with weights to the graph
G.add_weighted_edges_from([("A", "B", 1.5), ("B", "C", 2.0), ("C", "D", 2.5), ("D", "A", 1.0), ("A", "C", 3.0), ("C", "A", 3.0)])

# Draw the graph with arrows
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, arrows=True)

# Draw edge labels
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.show()
```
