````python
def flatlandSpaceStations(n, c):
    max_distance = 0
    for i in range(n):
        min_dist = float('inf')
        for station in c:
                distance = abs(i - station)
                min_dist = min(min_dist,distance)
        max_distance = max(max_distance, min_dist)

    return max_distance


print(flatlandSpaceStations(10, [2, 6, 8]))           
````

````python
def flatlandSpaceStations(n, c):
    c.sort()
    left_station = c[0] 
    right_station = c[-1]
    max_dist = max(left_station,(n-1)-right_station)
 
    for i in range(1,len(c)):
        distance = (c[i] - c[i-1])//2
        max_dist = max(max_dist, distance)
    return  max_dist


print(flatlandSpaceStations(10, [2, 6, 8]))
print(flatlandSpaceStations(15, [1, 4, 10, 13]))
````
