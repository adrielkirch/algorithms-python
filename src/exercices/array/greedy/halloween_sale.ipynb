
# ðŸŽ® Halloween Sale â€” Problem Statement

You wish to buy video games from the famous online store **Mist**.

Normally, all games cost **p dollars**.
For the **Halloween Sale**, prices follow this rule:

* The **first game** costs `p`.
* Each subsequent game costs **d dollars less** than the previous one.
* Once the price drops to **m or below**, every next game costs **m**.

Given a starting budget `s`, determine **how many games you can buy**.

---

## ðŸ§® Function Description

### `howManyGames(p, d, m, s)`

**Parameters:**

* `int p`: initial price of the first game
* `int d`: discount applied to each subsequent game
* `int m`: minimum price a game can reach
* `int s`: your starting budget

**Return value:**

* `int`: the **number of games** you can buy

---

## ðŸ“¥ Input Format

A single line containing four integers:

```
p d m s
```

---

## ðŸ“Œ Constraints

* `1 â‰¤ m â‰¤ p â‰¤ 100`
* `1 â‰¤ d â‰¤ 100`
* `1 â‰¤ s â‰¤ 10^4`

---

## ðŸ“˜ Sample Input 0

```
20 3 6 80
```

## ðŸ“— Sample Output 0

```
6
```

### Explanation

Game prices in order:

```
20, 17, 14, 11, 8, 6, 6, 6...
```

With a budget of **80**, you can buy:

```
20 + 17 + 14 + 11 + 8 + 6 = 76
```

That's **6 games**.
A seventh game would cost **6 more**, which exceeds 80.

---

## ðŸ“˜ Sample Input 1

```
20 3 6 85
```

## ðŸ“— Sample Output 1

```
7
```

### Explanation

With **85**, you can buy:

```
20 + 17 + 14 + 11 + 8 + 6 + 6 = 82
```

That's **7 games**.
An eighth would exceed the budget.



````python
def howManyGames(p, d, m, s):
    total = 0
    price = p
    while s >= price:
        s = s - price
        total +=1
        price = max(price - d, m)
    return total
````

````python
tests = [
    # 1. Sample Test 0
    {
        "p": 20, "d": 3, "m": 6, "s": 80,
        "expected": 6
    },
    # 2. Sample Test 1
    {
        "p": 20, "d": 3, "m": 6, "s": 85,
        "expected": 7
    },
    # 3. Price never hits minimum (small budget)
    {
        "p": 10, "d": 2, "m": 1, "s": 12,
        "expected": 1
    },
    # 4. Budget buys many minimum-price games
    {
        "p": 10, "d": 2, "m": 5, "s": 35,
        "expected": 5  # prices: 10,8,6,5,5
    },
    # 5. Only enough money for one game
    {
        "p": 50, "d": 5, "m": 20, "s": 50,
        "expected": 1
    },
    # 6. Cannot buy any game
    {
        "p": 30, "d": 5, "m": 5, "s": 10,
        "expected": 0
    },
    # 7. No discount at all
    {
        "p": 10, "d": 0, "m": 1, "s": 50,
        "expected": 5  # always pays 10
    },
    # 8. All prices already at minimum
    {
        "p": 5, "d": 10, "m": 5, "s": 21,
        "expected": 4  # 5,5,5,5
    }
]

for i, t in enumerate(tests, 1):
    result = howManyGames(t["p"], t["d"], t["m"], t["s"])
    print(f"Test {i}: got={result}, expected={t['expected']} â†’ "
          f"{'OK' if result == t['expected'] else 'FAIL'}")
````
