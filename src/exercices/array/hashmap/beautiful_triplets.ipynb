# ğŸ§® Beautiful Triplets â€” Problem Statement

Given an **increasing sequence of integers** and a value **d**, a **triplet** ((a[i], a[j], a[k])) is considered **beautiful** if these conditions hold:


a[j] - a[i] = d and a[k] - a[j] = d


In other words:


a[i], a[i] + d, a[i] + 2d


must all exist in the array.

Your task is to **count how many beautiful triplets** exist in the array.

---

## ğŸ”§ Function Description

Complete the function:

### `beautifulTriplets(d, arr)`

**Parameters:**

* `int d` â†’ the beautiful difference
* `int arr[n]` â†’ a sorted array of integers (ascending order)

**Returns:**

* `int` â†’ the number of beautiful triplets found in the array

---

## ğŸ“¥ Input Format

1. The first line contains two integers:

   * `n` â€” the size of the array
   * `d` â€” the beautiful difference
2. The second line contains `n` space-separated integers representing the array.

---

## ğŸ“ Constraints

* The array is guaranteed to be sorted in **ascending** order.

---

## ğŸ§ª Sample Input

```

```

## ğŸ“¤ Sample Output

```
3
```

## ğŸ“ Explanation

The beautiful triplets found are:

* ( (1, 4, 7) )
* ( (4, 7, 10) )
* ( (2, 5, 8) )

Each satisfies:


a[j] - a[i] = d and a[k] - a[j] = d


Therefore, the answer is **3**.

---

Let me know if you want the **Python solution**, **optimized version**, or **brute force vs. hash map explanation**!


````python
def beautifulTriplets(d, arr):
    seen = set(arr)
    count = 0
    for num in arr:
        if num + d  in seen and  num + 2*d  in seen:
            count += 1
    return  count
````
