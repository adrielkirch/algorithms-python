# Minimum Distances

## ğŸ“˜ Problem

Given an array of integers, the **distance** between two values is defined as the number of indices between them.

We must find the **minimum distance** between any pair of equal elements.
If no value appears more than once, return **-1**.

---

## ğŸ“Œ Example

### **Input**

```
6
7 1 3 4 1 7
```

### **Matching pairs**

| Value | Indices | Distance  |
| ----- | ------- | --------- |
| 1     | 1 and 4 | 4 - 1 = 3 |
| 7     | 0 and 5 | 5 - 0 = 5 |

### âœ” Minimum distance = **3**

---

## ğŸ§  Approach

Use a **hash map** (dictionary) to store the **last index** where each value was seen.

Steps:

1. Create a dict `last_seen`.
2. For each index `i`, check if the value has been seen before.
3. If yes:

   * compute the distance
   * update `min_dist`
4. Update `last_seen[value] = i`.
5. If `min_dist` was never updated, return `-1`.

### â± Time Complexity: `O(n)`

### ğŸ’¾ Space Complexity: `O(n)`

---

## ğŸ§ª Sample Run

Input:

```
7 1 3 4 1 7
```

Output:

```
3
```

````python
def minimumDistances(a):
    # Write your code here
    dict_index_map = {

    }
    min_dist = float('inf')
    for  i, num in enumerate(a):
        if a[i] in dict_index_map:
            dist = i - dict_index_map[num]
            min_dist = min(min_dist, dist)
        dict_index_map[num] = i
    return  min_dist if min_dist <  float('inf') else -1
````
