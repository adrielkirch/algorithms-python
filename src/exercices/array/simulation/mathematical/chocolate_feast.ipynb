# üç´ Chocolate Feast ‚Äî HackerRank Exercise

Little Bobby loves chocolate. His favorite store, **Penny Auntie**, is running a promotion:

- Every chocolate bar costs **c**.
- Every chocolate bar gives **1 wrapper**.
- By collecting **m wrappers**, Bobby can exchange them for **1 free chocolate**.

Your task:  
Given Bobby‚Äôs money **n**, chocolate cost **c**, and wrapper-exchange value **m**, determine **how many chocolates Bobby can eat** in total.

---

## üßÆ Function Description

### `chocolateFeast(n, c, m)`

| Parameter | Type | Description |
|----------|------|-------------|
| `n` | int | Bobby's initial amount of money |
| `c` | int | Cost of one chocolate bar |
| `m` | int | Number of wrappers to exchange for a free bar |

### **Returns**
`int` ‚Äî Total number of chocolates Bobby can eat.

---

## üìù Notes
- Bobby **always** exchanges wrappers whenever possible.
- The process continues until wrappers < m.

---

## üî¢ Example
**Input:**  
```

n = 10, c = 2, m = 5

```

**Process:**  
- Bobby buys `10 // 2 = 5` chocolates ‚Üí 5 wrappers  
- Exchange 5 wrappers ‚Üí +1 chocolate  
- Now wrappers = 1  
- No more exchanges  

**Total chocolates:** **6**

````python
def chocolateFeast (n, c, m):
    # How many chocolates can buy with the money?
    chocolates = n // c
    # Foreach choco 1 wrapper
    wrappers = chocolates
    # While the total of wrappers is more than m do
    while wrappers >= m:
        # How many free can get from the wrappers
        free = wrappers // m
        # total chocolates is sum wth free
        chocolates += free
        # Now we need to see how many wrappers last (wrappers % m) plus the free ones
        wrappers = wrappers % m + free
    return chocolates

    
````

````python
tests = [
    (10, 2, 5),
    (12, 4, 4),
    (6, 2, 2)
]

for n, c, m in tests:
    print(chocolateFeast(n, c, m))
````
