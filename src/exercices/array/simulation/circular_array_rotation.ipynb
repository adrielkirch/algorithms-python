# Circular Array Rotation

John Watson knows of an operation called a right circular rotation on an array of integers. One rotation operation moves the last array element to the first position and shifts all remaining elements right one. To test Sherlock's abilities, Watson provides Sherlock with an array of integers. Sherlock is to perform the rotation operation a number of times then determine the value of the element at a given position.

For each array, perform a number of right circular rotations and return the values of the elements at the given indices.

## Example

```
a = [3, 4, 5]
k = 2
queries = [1, 2]
```

Here `k` is the number of rotations on `a`, and `queries` holds the list of indices to report. First we perform the two rotations:

```
[3, 4, 5] -> [5, 3, 4] -> [4, 5, 3]
```

Now return the values from the zero-based indices `1` and `2` as indicated in the `queries` array.

```
a[1] = 5
a[2] = 3
```

## Function Description

Complete the circularArrayRotation function in the editor below.

circularArrayRotation has the following parameter(s):

- **int a[n]**: the array to rotate
- **int k**: the rotation count
- **int queries[q]**: the indices to report

## Returns

- **int[q]**: the values in the rotated array as requested in queries

## Input Format

The first line contains 3 space-separated integers, `n`, `k`, and `q`, the number of elements in the integer array, the rotation count and the number of queries.

The second line contains `n` space-separated integers, where each integer `i` describes array element `a[i]` (where `0 <= i < n`).

Each of the `q` subsequent lines contains a single integer, `queries[i]`, an index of an element in `a` to return.

## Constraints

- 1 <= n <= 10^5
- 1 <= a[i] <= 10^5
- 1 <= k <= 10^5
- 1 <= q <= 500
- 0 <= queries[i] < n

## Sample Input 0

```
3 2 3
1 2 3
0
1
2
```

## Sample Output 0

```
2
3
1
```

## Explanation 0

After the first rotation, the array is `[3, 1, 2]`.

After the second (and final) rotation, the array is `[2, 3, 1]`.

We will call this final state array `b`. For each query, we just have to get the value of `b[queries[i]]`.

- `queries[0] = 0`, `b[0] = 2`.
- `queries[1] = 1`, `b[1] = 3`.
- `queries[2] = 2`, `b[2] = 1`.

````python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'circularArrayRotation' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY a
#  2. INTEGER k
#  3. INTEGER_ARRAY queries
#

def rotate_once(a):
    return a[-1:] + a[:-1]


def circularArrayRotation(a, k, queries):
    for _ in range(k):
        a = rotate_once(a)
    return [a[x] for x in queries]

print(circularArrayRotation([3,4,5],2,[1,2]))
````

````python
def rotate_once(a):
    return a[-1:] + a[:-1]

def circularArrayRotation_best(a, k, queries):
    k = k % len(a)
    for _ in range(k):
        a = rotate_once(a)
    return [a[x] for x in queries]

# Test with the examples
print(circularArrayRotation_best([3, 4, 5], 2, [1, 2]))  # Expected: [5, 3]
print(circularArrayRotation_best([1, 2, 3], 2, [0, 1, 2]))  # Expected: [2, 3, 1]
````

## Solution 3: Best - Direct Index Calculation (No Array Rotation)

This is the most efficient solution. Instead of rotating the array, we calculate where each queried element would be after k rotations.

**Key insight**: After k right rotations, the element at index `i` in the rotated array comes from index `(i - k) % n` in the original array.

Time complexity: O(q) where q is the number of queries  
Space complexity: O(1) (excluding the result array)

````python
def circularArrayRotation_optimized(a, k, queries):
    # Optimize rotations using modular arithmetic
    n = len(a)
    k = k % n  # Only need to rotate k % n times
    
    # After k right rotations, rotate the array once efficiently
    """
    heuristic to visualize it
    k = 2
    [1,2,3,4] 
    [2,3,4,1] 1 rotation
    [3,4,1,2] 2 rotation

    a[-2:] [3,4]
    + 
    a[:-2] [1.2]
    [3,4,1,2]
    """
    rotated = a[-k:] + a[:-k] if k > 0 else a
    
    # Return values at queried indices
    return [rotated[query] for query in queries]

# Test with the examples
print(circularArrayRotation_optimized([3, 4, 5], 2, [1, 2]))  # Expected: [5, 3]
print(circularArrayRotation_optimized([1, 2, 3], 2, [0, 1, 2]))  # Expected: [2, 3, 1]
````

## Solution 2: Optimized with Modular Arithmetic (Rotate Array Once)

This solution uses modular arithmetic to optimize rotations. Instead of rotating k times, we:
1. Use `k % n` to reduce unnecessary rotations
2. Perform a single array slice operation

Time complexity: O(n)  
Space complexity: O(n)
