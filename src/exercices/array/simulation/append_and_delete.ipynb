## ğŸ§  Key Insight

The optimal approach is:

1. Find the **longest common prefix** between `s` and `t`
2. Count:

   * Characters to delete from the end of `s`
   * Characters to append to match `t`

Let:

```
common_prefix_length = length of longest common starting substring of s and t
operations_needed = (len(s) - common_prefix_length) + (len(t) - common_prefix_length)
```

Then the decision:
âœ” If we can match exactly `k` operations
âœ” Or we have **extra operations** that can be used by deleting from an empty string
â†’ Return **Yes**

Otherwise â†’ **No**

## ğŸ§ª Sample Inputs & Outputs

`````python
def find_common_prefix(s,t):
    common = 0
    for i in range(min(len(s),len(t))):
        if s[i] == t[i]:
            common += 1
        else:
            break
    return common

def appendAndDelete(s, t, k):
    common = find_common_prefix(s,t)
    total_operations = (len(s) - common) + (len(t) - common)
    if total_operations > k:
        return 'No'
    elif (k - total_operations) % 2 == 0:
        return 'Yes'
    elif len(s) + len(t) <= k:
        return 'Yes'
    return 'No'
`````

`````python
tests = [
    # Official HackerRank samples
    ("hackerhappy", "hackerrank", 9, "Yes"),
    ("aba", "aba", 7, "Yes"),
    ("ashley", "ash", 2, "No"),

    # Same prefix but extra operations must be handled
    ("aaa", "a", 5, "Yes"),  # deletes: 2, leftover 3 â‰ˆ odd â†’ but can delete empty

    # Completely different strings
    ("abc", "def", 6, "Yes"),
    ("abc", "def", 3, "No"),

    # k is too small even with minimal operations
    ("abcd", "abc", 0, "No"),
    ("abcd", "abc", 1, "Yes"),

    # Strings already equal but k is odd â†’ still possible
    ("aaa", "aaa", 1, "Yes"),
    ("aaa", "aaa", 2, "Yes"),
    
    # Large enough k to delete entire string and rebuild
    ("abcd", "xyz", 10, "Yes"),

    # Prefix differs in middle
    ("qwerasdf", "qwerbsdf", 6, "No"),

    # Edge cases with empty string
    ("", "abc", 3, "Yes"),
    ("abc", "", 3, "Yes"),
    ("", "", 5, "Yes"),  # delete on empty + re-add still ok
]
for s, t, k, expected in tests:
    result = appendAndDelete(s, t, k)
    print(f"s='{s}', t='{t}', k={k} => {result} | Expected: {expected}")
`````
