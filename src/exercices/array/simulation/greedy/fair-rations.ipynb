# Desafio: Fair Rations (Distribui√ß√£o Justa de P√£es)

## üìù Statement
Voc√™ √© o governante do Castelo Rankhacker e deve distribuir p√£es aos seus s√∫ditos que est√£o em uma fila. Alguns j√° possuem p√£es, mas voc√™ deve garantir que, ao final da distribui√ß√£o, **todos tenham uma quantidade par**.

### Regras de Distribui√ß√£o:
1. Ao dar um p√£o a uma pessoa, voc√™ **obrigatoriamente** deve dar um p√£o √† pessoa imediatamente √† frente ou atr√°s dela na fila.
2. O objetivo √© distribuir o **m√≠nimo poss√≠vel** de p√£es.
3. Se n√£o for poss√≠vel deixar todos com uma contagem par, retorne **"NO"**.

### Exemplo:
Fila inicial: `[2, 3, 4, 5, 6]`
1. Damos 1 p√£o ao segundo e ao terceiro: `[2, 4, 5, 5, 6]`
2. Damos 1 p√£o ao terceiro e ao quarto: `[2, 4, 6, 6, 6]`
**Total distribu√≠do: 4 p√£es.**

---

## üèóÔ∏è Classifica√ß√£o do Problema
Este √© um problema cl√°ssico de **Algoritmo Guloso (Greedy Algorithm)** com foco em **Paridade**.

* **Tipo:** Otimiza√ß√£o / Greedy.
* **L√≥gica:** Percorremos a fila uma √∫nica vez. Sempre que encontramos um s√∫dito com p√£es √≠mpares, somos for√ßados a dar um p√£o a ele e ao pr√≥ximo. Essa decis√£o √© localmente necess√°ria e leva √† solu√ß√£o m√≠nima global.
* **Complexidade de Tempo:** $O(N)$ - Uma √∫nica passagem pelo array.
* **Complexidade de Espa√ßo:** $O(1)$ - Apenas vari√°veis de contagem.

---

## üõ†Ô∏è Heur√≠stica de Resolu√ß√£o
A l√≥gica baseia-se na **propaga√ß√£o da imparidade**:
1. Verificamos se a soma total de p√£es √© √≠mpar (se for, √© imposs√≠vel resolver, pois cada a√ß√£o adiciona 2 p√£es).
2. Varremos o array da esquerda para a direita.
3. Se o elemento atual for √≠mpar, somamos +1 nele e no pr√≥ximo, incrementando o contador total em 2.
4. Ao final, a "batata quente" da imparidade ter√° sido empurrada at√© encontrar outro √≠mpar e se anular.

Vamos analisar um exemplo pequeno e ver os padr√µes

**B = [1,2]**

N√£o da porque sempre vai ficar impar

**B = [1,1] **

Vai dar certo, porque de +1 ambos ficam par

**B = [1,2,3]**
**B = [2,3,3]**
**B = [2,4,4]**

Vai dar certo chegamos

**B = [2,3,4]**
**B = [2,4,5]**
**B = [3,5,5]**
**B = [4,6,5]**
**B = [4,7,6]** 

voltou pro mesmo padr√£o inicial P,I,P

Se olharemos os exemplos a soma dos que deram certo √© um numero par e os que n√£o deram √© impar

1. O Princ√≠pio da Obrigatoriedade (O "N√≥" na Fila)Imagine a primeira pessoa da fila (B[0]).

Se ela tem um n√∫mero par de p√£es, voc√™ n√£o deve dar p√£o a ela. Por qu√™? 

Porque se der, ela ficar√° √≠mpar, e para consertar isso voc√™ ter√° que dar outro p√£o a ela mais tarde, gastando mais do que o m√≠nimo.Se ela tem um n√∫mero √≠mpar de p√£es, voc√™ √© obrigado a dar um p√£o a ela. 

Como n√£o existe ningu√©m antes dela (i-1), a √∫nica forma de torn√°-la par seguindo as regras √© dando um p√£o para ela e para o vizinho da direita (i+1).Nesse momento, o estado da primeira pessoa est√° selado. 

Ela est√° par e voc√™ n√£o voltar√° mais nela.

2. A Rea√ß√£o em Cadeia


Agora olhe para a segunda pessoa (B[1]). O n√∫mero de p√£es dela pode ter mudado porque ela recebeu um p√£o da primeira pessoa.Novamente, se ela estiver √≠mpar agora, a √∫nica maneira de torn√°-la par sem bagun√ßar a primeira pessoa (que j√° selamos) √© dando um p√£o para ela e para o pr√≥ximo (B[2]).

```python
def fairRations(B):
    total_sum = sum(x for x in B)
    total_operations = 0

    if total_sum  % 2 == 1:
        return "NO"

    for i in range(len(B) - 1):
        if B[i] % 2 == 1:
            B[i] += 1
            B[i+1] += 1
            total_operations += 2
    return str(total_operations)
```

```python
test_cases = [
    {
        "nome": "Exemplo do Enunciado",
        "entrada": [2, 3, 4, 5, 6],
        "esperado": "4"
    },
    {
        "nome": "Caso Imposs√≠vel (Soma √çmpar)",
        "entrada": [1, 2],
        "esperado": "NO"
    },
    {
        "nome": "Padr√£o I-P-I (Seu exemplo)",
        "entrada": [1, 2, 1],
        "esperado": "4"
    },
    {
        "nome": "Todos j√° s√£o Pares",
        "entrada": [2, 4, 10, 8],
        "esperado": "0"
    },
    {
        "nome": "√çmpares nas extremidades",
        "entrada": [1, 2, 2, 2, 1],
        "esperado": "8" 
        # (O 1 da ponta vai empurrando a batata quente at√© o outro 1)
    },
    {
        "nome": "Dois √≠mpares juntos",
        "entrada": [2, 3, 5, 6],
        "esperado": "2"
    }
]

for case in test_cases:
    resultado = fairRations(case["entrada"])
    status = "‚úÖ PASSOU" if resultado == case["esperado"] else "‚ùå FALHOU"
    print(f"{case['nome']:<30} | {resultado:<10} | {status}")
```
