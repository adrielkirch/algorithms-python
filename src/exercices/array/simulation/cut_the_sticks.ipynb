# Cut the Sticks

## Problem Description

You are given a number of sticks of varying lengths. You will iteratively cut the sticks into smaller sticks, discarding the shortest pieces until there are none left. At each iteration you will determine the length of the shortest stick remaining, cut that length from each of the longer sticks and then discard all the pieces of that shortest length. When all the remaining sticks are the same length, they cannot be shortened so discard them.

Given the lengths of `n` sticks, print the number of sticks that are left before each iteration until there are none left.

## Example

`arr = [5, 4, 4, 2, 2, 8]`

The shortest stick length is `2`, so cut that length from the longer two and discard the pieces of length `2`. Now the lengths are `[3, 2, 2, 6]`. Again, the shortest stick is of length `2`, so cut that amount from the longer stick and discard those pieces. There is only one stick left, `4`, so discard that stick. The number of sticks at each iteration are `[6, 4, 2, 1]`.

## Function Description

Complete the `cutTheSticks` function in the editor below. It should return an array of integers representing the number of sticks before each cut operation is performed.

**cutTheSticks** has the following parameter(s):
- `int arr[n]`: the lengths of each stick

**Returns:**
- `int[]`: the number of sticks after each iteration

## Input Format

The first line contains a single integer `n`, the size of `arr`.
The next line contains `n` space-separated integers, each an `arr[i]`, where each value represents the length of the `i`th stick.

## Solution

````python
import math
import os
import random
import re
import sys

#
# Complete the 'cutTheSticks' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts INTEGER_ARRAY arr as parameter.
#

def cutTheSticks(arr):
    result = []
    
    while len(arr) > 0:
        # Record the number of sticks before this iteration
        result.append(len(arr))
        
        # Find the minimum stick length
        min_stick = min(arr)
        
        # Cut all sticks by the minimum length and remove sticks of length 0
        arr = [stick - min_stick for stick in arr if stick - min_stick > 0 ]
    
    return result
````

````python
# ðŸ”¥ Key heuristic:
# Everything from index 0 to i-1 is "gone".
# Everything from index i to n-1 is still > 0.
#
# So the number of sticks remaining is:
#   n - i
#
# Why does this work?
# Because the array is sorted:
#   - All smaller elements are grouped at the start.
#   - Once we reach a new value, all earlier sticks
#     would have been cut down to zero.
#
# Example:
# arr = [2,2,4,4,5,8]
# i = 2  (first place where 2 â†’ 4 changes)
# arr[2:] = [4,4,5,8] â†’ 4 sticks remain
# n - i = 6 - 2 = 4
def cutTheSticks(arr):
    # Sort once so all equal lengths stay together
    # and increasing groups come in order.
    arr.sort()
    
    result = []
    n = len(arr)

    # Before any "cuts", all sticks exist
    result.append(n)

    # Walk through the sorted array
    for i in range(1, n):

        # Whenever we see a new value (a jump),
        # it means all sticks before index 'i' had a smaller length.
        # Those are the sticks that would become 0 after cutting.
        if arr[i] != arr[i - 1]:

          
            remaining = n - i

            result.append(remaining)

    return result
````
