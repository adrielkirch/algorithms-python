# Mini-Max Sum Exercise

## Problem Description

Given five positive integers, find the minimum and maximum values that can be calculated by summing exactly four of the five integers. Then print the respective minimum and maximum values as a single line of two space-separated long integers.

## Example

`arr = [7, 69, 2, 221, 8974]`

The minimum sum is obtained by taking all numbers except the largest: 7 + 69 + 2 + 221 = 299
The maximum sum is obtained by taking all numbers except the smallest: 69 + 2 + 221 + 8974 = 9266

The function should print:
```
299 9266
```

## Function Description

Complete the `miniMaxSum` function with the following parameter(s):

* `arr[5]`: an array of 5 positive integers

## Print

Print two space-separated integers on one line: the minimum sum and the maximum sum of 4 of 5 elements.

## Note
For some languages, like C, C++, and Java, the sums may require that you use a long integer due to their size.

## Input Format

A single line of five space-separated integers.

## Constraints

* 1 ≤ arr[i] ≤ 10^9

## Sample Input

```
7 69 2 221 8974
```

## Expected Output

```
299 9266
```

````python
#!/bin/python3

import math
import os
import random
import re
import sys

def miniMaxSum(arr):
    """
    Find and print the minimum and maximum sum possible by using 4 of 5 integers
    
    Args:
        arr (list): An array of 5 positive integers
        
    Returns:
        None: This function prints the minimum and maximum sum
    """
    arr.sort()
    min_sum = sum(arr[:-1])
    max_sum = sum(arr[1:])
    print(f"{min_sum} {max_sum}")
    pass

# Test with the given example
test_arr = [7, 69, 2, 221, 8974]
print(f"Testing with: {test_arr}")
miniMaxSum(test_arr)

# Test with another example
test_arr2 = [1, 3, 5, 7, 9]
print(f"\nTesting with: {test_arr2}")
miniMaxSum(test_arr2)
````

````python
#!/bin/python3

import math
import os
import random
import re
import sys

def miniMaxSum(arr):
    """
    sum -> returns sum in an array
    min -> returns the minimal in an array 
    max -> returns the max in an array

    For min -> we sum all values but not sum the max
    For max -> we sum all values but not sum the min

    We could sum everything and subtract the undesired value
    """
    total = sum(arr)
    min_value = min(arr)
    max_value = max(arr)
    min_sum = total - max_value
    max_sum = total - min_value
    print(f"{max_sum} {min_sum}")
    pass

# Test with the given example
test_arr = [7, 69, 2, 221, 8974]
print(f"Testing with: {test_arr}")
miniMaxSum(test_arr)

# Test with another example
test_arr2 = [1, 3, 5, 7, 9]
print(f"\nTesting with: {test_arr2}")
miniMaxSum(test_arr2)
````
