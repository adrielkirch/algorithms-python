
#  Problema: Kangaroo

Dois cangurus esto em uma linha, cada um comeando em uma posio diferente, e pulam em direo ao infinito positivo.  
Devemos determinar se eles se encontraro na mesma posio aps algum nmero de pulos.

**Parmetros:**
- `x1, v1`  posio inicial e tamanho do salto do canguru 1  
- `x2, v2`  posio inicial e tamanho do salto do canguru 2  

**Retorno:**
- `"YES"` se eles se encontrarem  
- `"NO"` caso contrrio  

**Exemplo:**

```

x1 = 0, v1 = 3
x2 = 4, v2 = 2

```

Aps um salto:  
- Canguru 1: 0 + 3 = 3  
- Canguru 2: 4 + 2 = 6  

Aps dois saltos:  
- Canguru 1: 6  
- Canguru 2: 8  

Aps trs saltos:  
- Canguru 1: 9  
- Canguru 2: 10  

No se encontram?  **NO**

````python
def kangaroo(x1, v1, x2, v2):
    
    if x1 == x2:
        return "YES"

    if v1 == v2:
        #return "YES" if x1 == x2 else "NO"
        return "NO"
    
    if v2 > v1:
        return "NO"
   
    if (x2 - x1) % (v1 - v2) == 0:
        return "YES"

    return "NO"
````

````python
#  Testes
test_cases = [
    (0, 3, 4, 2, "YES"),
    (0, 2, 5, 3, "NO"),
    (0, 2, 0, 2, "YES"),
    (0, 2, 1, 2, "NO"),
]

for i, (x1, v1, x2, v2, esperado) in enumerate(test_cases, 1):
    result = kangaroo(x1, v1, x2, v2)
    status = " PASSOU" if result == esperado else f" FALHOU (esperado {esperado}, obtido {result})"
    print(f" Teste {i}: x1={x1}, v1={v1}, x2={x2}, v2={v2}  {result} {status}")
````
