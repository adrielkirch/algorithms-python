# Day of Programmer

## Problem Description

Marie invented a Time Machine and wants to test it by time-traveling to visit Russia on the Day of the Programmer (the 256th day of the year) during a year in the inclusive range from 1700 to 2700.

From 1700 to 1917, Russia's official calendar was the Julian calendar; since 1919 they used the Gregorian calendar system. The transition from the Julian to Gregorian calendar system occurred in 1918, when the next day after January 31st was February 14th. This means that in 1918, February 14th was the 32nd day of the year in Russia.

In both calendar systems, February is the only month with a variable amount of days; it has 29 days during a leap year, and 28 days during all other years. In the Julian calendar, leap years are divisible by 4; in the Gregorian calendar, leap years are either of the following:

- Divisible by 400.
- Divisible by 4 and not divisible by 100.

Given a year, find the date of the 256th day of that year according to the official Russian calendar during that year. Then print it in the format `dd.mm.yyyy`, where `dd` is the two-digit day, `mm` is the two-digit month, and `yyyy` is the year.

For example, the given year = 1984. 1984 is divisible by 4, so it is a leap year. The 256th day of a leap year after 1918 is September 12, so the answer is `12.09.1984`.

## Function Description

Complete the `dayOfProgrammer` function in the editor below. It should return a string representing the date of the 256th day of the year given.

`dayOfProgrammer` has the following parameter(s):
- `year`: an integer

## Input Format

A single integer denoting year.

## Constraints

## Output Format

Print the full date of Day of the Programmer during year in the format `dd.mm.yyyy`, where `dd` is the two-digit day, `mm` is the two-digit month, and `yyyy` is the year.

## Sample Input 0

```

## Sample Output 0

```

## Explanation 0

In the year 2017, January has 31 days, February has 28 days, March has 31 days, April has 30 days, May has 31 days, June has 30 days, July has 31 days, and August has 31 days. When we sum the total number of days in the first eight months, we get 31 + 28 + 31 + 30 + 31 + 30 + 31 + 31 = 243. Day of the Programmer is the 256th day, so then calculate 256 - 243 = 13 to determine that it falls on day 13 of the 9th month (September). We then print the full date in the specified format, which is 13.09.2017.

We then print the full date in the specified format, which is 12.09.2016.

## Key Points to Consider

1. **Calendar Systems:**
   - Julian calendar: 1700-1917
   - Transition year: 1918 (special case)
   - Gregorian calendar: 1919-2700

2. **Leap Year Rules:**
   - Julian: divisible by 4
   - Gregorian: divisible by 400 OR (divisible by 4 AND not divisible by 100)

3. **Special Case 1918:**
   - January 31st was followed by February 14th
   - February 14th was the 32nd day of the year

4. **Target Day:**
   - Find the 256th day of the year
   - Usually falls in September

````python
def dayOfProgrammer(year):
    """
    Find the date of the 256th day of the year according to the Russian calendar.
    
    Args:
        year: Integer representing the year (1700 <= year <= 2700)
        
    Returns:
        String in format "dd.mm.yyyy" representing the 256th day
    """
    TARGET_DAY = 256

    sum_days = 0 

    is_leap = is_leap_year(year)
    dict_month_days = get_month_total_days(year,is_leap)

    for month, total_days in dict_month_days.items():
        if sum_days + total_days >= TARGET_DAY:
            days_into_month = TARGET_DAY - sum_days
            return f'{days_into_month:02d}.{month}.{year}'
        sum_days += total_days
        

def is_leap_year(year):
    """
    Determine if a year is a leap year based on the calendar system.
    
    Args:
        year: Integer representing the year
        
    Returns:
        Boolean indicating if the year is a leap year
    """
    if year <= 1917:
        return year % 4 == 0
    if year >= 1919:
         return year % 400 == 0 or (year % 4 == 0 and year % 100 > 0)
    else:
        return False

def get_month_total_days(year, is_leap=False):
    days_year = {
        '01':31,
        '02':28,
        '03':31,
        '04':30,
        '05':31,
        '06':30,
        '07':31,
        '08':31,
        '09':30,
        '10':31,
        '11':30,
        '12':31,
    }
    if is_leap:
        days_year['02'] = 29
    elif year == 1918:
        days_year['02'] = 15
    return days_year
        

 # Test cases
if __name__ == "__main__":
    # Test case 1: Normal Gregorian year
    test1 = 2017
    expected1 = "13.09.2017"
    result1 = dayOfProgrammer(test1)
    print(f"Test 1: {test1} -> Expected: {expected1}, Got: {result1}")
    
    # Test case 2: Gregorian leap year
    test2 = 2016
    expected2 = "12.09.2016"
    result2 = dayOfProgrammer(test2)
    print(f"Test 2: {test2} -> Expected: {expected2}, Got: {result2}")
    
    # Test case 3: Julian calendar year
    test3 = 1800
    expected3 = "12.09.1800"
    result3 = dayOfProgrammer(test3)
    print(f"Test 3: {test3} -> Expected: {expected3}, Got: {result3}")
    
    # Test case 4: Transition year (special case)
    test4 = 1918
    expected4 = "26.09.1918"  # This needs to be calculated considering the transition
    result4 = dayOfProgrammer(test4)
    print(f"Test 4: {test4} -> Expected: {expected4}, Got: {result4}")
    
    # Additional test cases
    test5 = 2000  # Gregorian leap year (divisible by 400)
    result5 = dayOfProgrammer(test5)
    print(f"Test 5: {test5} -> Got: {result5}")
    
    test6 = 1900  # Not a leap year in Gregorian (divisible by 100 but not 400)
    result6 = dayOfProgrammer(test6)
    print(f"Test 6: {test6} -> Got: {result6}")   
````

## Solution Approach

Think about these steps:

1. **Identify the calendar system** based on the year
2. **Determine if it's a leap year** using the appropriate rules
3. **Calculate days in each month** up to August (month 8)
4. **Handle the special case of 1918** where February starts on the 14th
5. **Calculate the remaining days** needed to reach the 256th day
6. **Format the result** as "dd.mm.yyyy"
