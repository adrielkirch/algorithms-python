
# Manasa and Stones — Explicação do Problema

## 📘 Descrição do Problema

Manasa está seguindo uma trilha de pedras.
Cada pedra tem um número gravado nela.

Ela observa que:

* Ela **começa sempre na pedra 0**
* Existem **`n` pedras adicionais** após a pedra inicial
* A diferença entre duas pedras consecutivas **sempre é um dos dois valores possíveis**:

  * `a` **ou**
  * `b`

O objetivo é **descobrir todos os valores possíveis que a ÚLTIMA pedra pode ter**, considerando **todas as combinações possíveis** dessas diferenças.

---

## 🧠 O que exatamente precisa ser feito?

Dado:

* `n` → número de pedras **após a pedra 0**
* `a` → diferença possível entre pedras
* `b` → outra diferença possível

Você deve:

1. Considerar **todas as sequências possíveis** de `n` diferenças
2. Cada diferença pode ser `a` **ou** `b`
3. Somar essas diferenças começando do valor `0`
4. Guardar **todos os valores finais possíveis**
5. Remover duplicados
6. Retornar os valores **em ordem crescente**

---

## 📌 Regras Importantes (essas confundem!)

* A pedra inicial **0 NÃO conta** como uma das `n` pedras
* Você aplica **exatamente `n-1` diferenças**
* A **ordem das diferenças importa**
* Mas o **resultado final pode se repetir**, e duplicatas devem ser removidas

---

## ✍️ Exemplo Simples
Olhe para a tabela SEM pensar em matemática

Você vê isto:

Diferenças
1,1
1,2
2,1
2,2

Pergunta humana, não matemática:

❓ Quantos números existem em cada linha?

Resposta óbvia:
👉 2 números

Ou seja:

toda linha tem 2 passos

nunca 1

nunca 3

📌 Isso já diz algo muito importante:

O número de passos é FIXO

2️⃣ Agora ligue isso com n = 3

Você sabe que:

0 → pedra → pedra


Conte com o dedo:

do 0 até a primeira pedra → 1 passo

da primeira até a segunda → 1 passo

👉 Total: 2 passos

Isso bate exatamente com a tabela:

cada linha tem 2 diferenças

Então, sem fórmula nenhuma:

n - 1 = número de diferenças

### Entrada

```
n = 3
a = 1
b = 2
```

### Possíveis combinações de diferenças (2 diferenças no total):

| Diferenças | Caminho   | Última pedra |
| ---------- | --------- | ------------ |
| 1,1        | 0 → 1 → 2 | 2            |
| 1,2        | 0 → 1 → 3 | 3            |
| 2,1        | 0 → 2 → 3 | 3            |
| 2,2        | 0 → 2 → 4 | 4            |

### Resultado final

```
2 3 4
```

x + y = n - 1

---

## 🔍 Observação Crucial (insight do problema)

Não importa **a ordem exata** das diferenças, apenas:

* Quantas vezes `a` aparece
* Quantas vezes `b` aparece

Se:

* `x` vezes usamos `a`
* `y` vezes usamos `b`
* então `x + y = n - 1`

O valor final será:

```
x * a + y * b
```

## 🧮 Exemplo maior

### Entrada

```
n = 4
a = 10
b = 100
```

Aqui temos **3 diferenças** (`n-1`).

Possíveis valores finais:

```
30   (3*10)
120  (2*10 + 1*100)
210  (1*10 + 2*100)
300  (3*100)
```

### Saída

```
30 120 210 300
```

---

## 🏷️ Categoria do Problema

Este problema é classificado como:

### ✅ **Matemática + Observação / Padrões**

* Não é backtracking
* Não é força bruta
* Não é DP
* Não é greedy clássico

📌 O desafio está em **perceber o padrão matemático**, não em gerar todas as permutações.

---

## ⛔ Limitações Importantes

* `n` pode ser grande → gerar todas as permutações seria inviável
* Precisamos de uma solução **O(n)** no máximo
* Duplicatas devem ser evitadas

---

## 🧠 Resumo Mental (para prova / entrevista)

“A última pedra depende apenas de quantas vezes escolhemos `a` e `b`, não da ordem.”

````python
def stones(n, a, b):
    results = set()
    for i in range(n):
        # x + y = n - 1
        # y = n - 1 - x
        # a*x + b*y = valor_final
        # valor_final = a*x + b*(n - 1 - x)
        value = i*a + b*(n - 1 -i) 
        results.add(value)
    return sorted(results)
````
