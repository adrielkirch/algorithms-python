You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

Example 1:

 ![alt text](image-2.png)

Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
Example 2:

Input: list1 = [], list2 = []
Output: []
Example 3:

Input: list1 = [], list2 = [0]
Output: [0]

```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def mergeTwoLists(self, list1, list2):
        print("\n🚀 Iniciando mergeTwoLists()")
        print("Lista 1 inicial:", self.list_to_array(list1))
        print("Lista 2 inicial:", self.list_to_array(list2))
        print("-" * 40)

        merged_list = ListNode()
        current = merged_list
        step = 1

        while list1 and list2:
            print(f"\n🔹 Passo {step}:")
            print(f"  list1.val = {list1.val}, list2.val = {list2.val}")
            
            if list1.val <= list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next

            current = current.next
            step += 1

        # Conecta o restante da lista que sobrou
        current.next = list1 if list1 else list2

        print("\n✅ Lista final mesclada:", self.list_to_array(merged_list.next))
        print("=" * 40)
        return merged_list.next

    def print_list(self, head: Optional[ListNode]):
        current = head
        while current:
            print(current.val, end=" -> ")
            current = current.next
        print("None")

    def list_to_array(self, node: Optional[ListNode]):
        """Função auxiliar só para exibir a lista como array no log"""
        arr = []
        while node:
            arr.append(node.val)
            node = node.next
        return arr

l1 = ListNode(1, ListNode(2, ListNode(4)))
l2 = ListNode(1, ListNode(3, ListNode(4)))
sol = Solution()
merged = sol.mergeTwoLists(l1, l2)
print("\n🧾 Resultado final:")
sol.print_list(merged)
```
