

# 140. Word Break II

Given a string s and a dictionary of strings wordDict, add spaces in s to construct a sentence where each word is a valid dictionary word. Return all such possible sentences in any order.

Note that the same word in the dictionary may be reused multiple times in the segmentation.

 
Example 1:

Input: s = "catsanddog", wordDict = ["cat","cats","and","sand","dog"]
Output: ["cats and dog","cat sand dog"]
Example 2:

Input: s = "pineapplepenapple", wordDict = ["apple","pen","applepen","pine","pineapple"]
Output: ["pine apple pen apple","pineapple pen apple","pine applepen apple"]
Explanation: Note that you are allowed to reuse a dictionary word.
Example 3:

Input: s = "catsandog", wordDict = ["cats","dog","sand","and","cat"]
Output: []
 

Constraints:

1 <= s.length <= 20
1 <= wordDict.length <= 1000
1 <= wordDict[i].length <= 10
s and wordDict[i] consist of only lowercase English letters.

# 3️⃣ #Implementation

## 🧼 Versão limpa e didática (Top-Down DP)

```python
from typing import List, Dict

class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> List[str]:
        # Combine uma verificação DP (prune) com DFS+memo para ser didático e eficiente
        word_set = set(wordDict)  # O(1) lookup
        n = len(s)
        # 1) DP de viabilidade: dp[i] = True se s[:i] pode ser segmentada
        dp = [False] * (n + 1)
        dp[0] = True
        for i in range(1, n + 1):
            for j in range(i):
                if dp[j] and s[j:i] in word_set:
                    dp[i] = True
                    break
        if not dp[n]:
            return []  # pruning: nenhuma sentença possível

        # 2) DFS com memo explícito para construir frases
        memo: Dict[int, List[str]] = {}

        def dfs(start: int) -> List[str]:
            if start == n:
                return ['']  # base para concatenação
            if start in memo:
                return memo[start]

            res: List[str] = []
            # tenta cortes e usa dp[end] para podar sufixos impossíveis
            for end in range(start + 1, n + 1):
                if not dp[end]:
                    continue
                prefix = s[start:end]
                if prefix not in word_set:
                    continue
                sentences_end = dfs(end):
                for suffix in sentences_end:
                    if suffix == '':
                        res.append(prefix)
                    else:
                        res.append(prefix + ' ' + suffix)
            memo[start] = res
            return res

        return dfs(0)
```

---

## 🧠 Comentários importantes

* `return [""]` → hack elegante para evitar espaço extra
* `word_set` → evita O(N) lookup
* `memo[start]` → evita recomputação exponencial

# 🚀 Insight Final (nível senior)

Esse problema é:

````python
from typing import List, Dict

def word_break_all(s: str, word_dict: List[str]) -> List[str]:
    words = set(word_dict)
    n = len(s)

    # 1) DP de viabilidade: dp[i] = True se s[:i] pode ser segmentada
    dp = [False] * (n + 1)
    dp[0] = True
    for i in range(1, n + 1):
        for j in range(i):
            if dp[j] and s[j:i] in words:
                dp[i] = True
                break
    if not dp[n]:
        return []  # pruning: nenhuma sentença possível

    # 2) DFS com memo explícito para construir todas as frases
    memo: Dict[int, List[str]] = {}

    def dfs(start: int) -> List[str]:
        if start == n:
            return ['']
        if start in memo:
            return memo[start]

        res: List[str] = []
        for end in range(start + 1, n + 1):
            if not dp[end]:         # poda importante
                continue
            prefix = s[start:end]
            if prefix not in words:
                continue
            subsentences = dfs(end)
            for subsentence in subsentences:
                res.append(prefix if subsentence == '' else prefix + ' ' + subsentence)
        memo[start] = res
        return res

    return dfs(0)
````
