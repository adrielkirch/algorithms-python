# üçé Count Apples and Oranges

## üìò Descri√ß√£o

Sam possui uma casa localizada entre dois pontos em uma linha num√©rica:  
- **`s`** ‚Üí ponto inicial da casa  
- **`t`** ‚Üí ponto final da casa  

√Ä esquerda da casa existe uma **macieira** no ponto **`a`**, e √† direita, uma **laranjeira** no ponto **`b`**.  
Cada fruta cai a uma certa dist√¢ncia **`d`** do ponto da √°rvore:

- `d > 0` ‚Üí a fruta cai **√† direita** da √°rvore  
- `d < 0` ‚Üí a fruta cai **√† esquerda** da √°rvore  

O objetivo √© determinar **quantas ma√ß√£s e quantas laranjas caem dentro da √°rea da casa** (entre `s` e `t`, inclusive).

---



```text
Casa: s = 7, t = 10  
√Årvore de ma√ß√£: a = 4  
√Årvore de laranja: b = 12  
Ma√ß√£s lan√ßadas: [2, 3, -4] ‚Üí posi√ß√µes finais = [6, 7, 0]  
Laranjas lan√ßadas: [3, -2, -4] ‚Üí posi√ß√µes finais = [15, 10, 8]  

Resultado:
1 ma√ß√£ e 2 laranjas ca√≠ram na casa.


````python
"""
a -> apple tree, m total apples
b -> Orangre, n total oranges
s -> t House Area
When fruit fall from tree it has "d" units
d > 0 means it fall to the right
d < 0 means it fall to the left 
Given the value of d for m apples and n orages, how many apples and orages in house area s->t

ex:
    s = 7 and t = 10
    a = 4, m = 3, thrown = [2,3,-4], [4+2, 4+3, 4-4] = [6,7,0]
    b = 12, n = 3, thrown = [3,-2,-4], [12+3,12-2,12-4] = [15,10,8]
    
    so in the house it fall 1 apple and 2 orange

"""
def is_between(s,t,d):
    return d <= t and d >= s

def countApplesAndOranges(s, t, a, b, apples, oranges):

    thrown_dist_apples = [a + apple for apple in apples]
    thrown_dist_oranges = [b + orange for orange in oranges]

    total_apple_house = sum(1 for apple in thrown_dist_apples if is_between(s,t,apple))
    total_orange_house = sum(1 for orange in thrown_dist_oranges if is_between(s,t,orange))
    print(total_apple_house)
    print(total_orange_house)
    return total_apple_house, total_orange_house
    
````

````python

# ---------------------------------------
# üîç Testes autom√°ticos
tests = [
    # (s, t, a, b, apples, oranges, esperado)
    (7, 10, 4, 12, [2, 3, -4], [3, -2, -4], (1, 2)),
    (5, 12, 4, 12, [-1, 2, 4], [-2, 0, 3], (2, 2)),
    (10, 15, 5, 15, [1, 2, 5], [-3, -5, 0], (1, 3)),
    (0, 5, 4, 5, [-1,0,2], [-2,-1,1], (2, 2))
]

for i, (s, t, a, b, apples, oranges, esperado) in enumerate(tests, 1):
    result = countApplesAndOranges(s, t, a, b, apples, oranges)
    ok = "‚úÖ PASSOU" if result == esperado else f"‚ùå FALHOU (esperado {esperado}, obtido {result})"
    print(f"üîπ Teste {i}\nCasa: {s}‚Äì{t}, √Årvores: a={a}, b={b}")
    print(f"Ma√ß√£s: {apples}, Laranjas: {oranges}\n{ok}")
    print("-" * 50)
````
