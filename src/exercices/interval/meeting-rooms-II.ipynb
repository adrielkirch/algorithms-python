# Meeting Rooms II

## Problem Description
Given an array of meeting time intervals where each interval is [start, end], determine the minimum number of meeting rooms required.

## Example
```
Input: [[0,30], [5,10], [15,20]]
Output: 2
Explanation: We need at least 2 rooms. Room 1: [0,30], Room 2: [5,10] and [15,20]

Input: [[7,10], [2,4]]
Output: 1
```

## Complexity
- **Time**: O(n log n) - sorting
- **Space**: O(n) - heap

## Why Uber cares
- **Scheduling drivers**: Managing concurrent deliveries
- **Resource allocation**: Determining server capacity needed
- **Load balancing**: Peak traffic handling

## Approach 1: Min Heap (Optimal)

### Idea:
1. Sort intervals by start time
2. Use min-heap to track end times of ongoing meetings
3. For each meeting:
   - If earliest ending meeting ends before current starts, reuse room
   - Otherwise, need a new room
4. Heap size = rooms needed

### Example:
```
Intervals: [[0,30], [5,10], [15,20]]
Sorted: [[0,30], [5,10], [15,20]]

[0,30]: heap = [30], rooms = 1
[5,10]: 30 > 5? Yes → add, heap = [10, 30], rooms = 2
[15,20]: 10 < 15? Yes → pop 10, add 20, heap = [20, 30], rooms = 2

Result: 2 rooms
```

````python
import heapq
from typing import List

class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        """Min Heap approach - O(n log n) time, O(n) space"""
        if not intervals:
            return 0
        
        # Sort by start time
        intervals.sort()
        
        # Min heap of end times
        rooms = []
        
        for start, end in intervals:
            # If earliest room becomes free before current meeting starts
            if rooms and rooms[0] <= start:
                heapq.heapreplace(rooms, end)  # Pop min and push end
            else:
                heapq.heappush(rooms, end)  # Need a new room
            
            print(f"Meeting [{start}, {end}]: rooms = {sorted(rooms)}, count = {len(rooms)}")
        
        return len(rooms)

# Test
sol = Solution()
print(f"Result: {sol.minMeetingRooms([[0,30], [5,10], [15,20]])}\n")  # 2
print(f"Result: {sol.minMeetingRooms([[7,10], [2,4]])}\n")  # 1
print(f"Result: {sol.minMeetingRooms([[1,5], [1,5], [1,5]])}\n")  # 3
````

## Approach 2: Chronological Ordering (Alternative)

### Idea:
1. Separate start and end times
2. Sort both arrays
3. Use two pointers to track concurrent meetings

````python
class Solution2:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        """Chronological approach - O(n log n) time"""
        if not intervals:
            return 0
        
        starts = sorted([interval[0] for interval in intervals])
        ends = sorted([interval[1] for interval in intervals])
        
        rooms_needed = 0
        i, j = 0, 0
        
        while i < len(starts):
            # If next start < next end, need a new room
            if starts[i] < ends[j]:
                rooms_needed += 1
                i += 1
            else:
                # A room becomes free
                rooms_needed -= 1
                j += 1
        
        return rooms_needed

sol2 = Solution2()
print(f"Result: {sol2.minMeetingRooms([[0,30], [5,10], [15,20]])}\n")  # 2
````

## Key Insights

1. **Greedy Reuse**: Always reuse the room that frees up earliest (min-heap property)
2. **Interval Overlap**: Two intervals overlap if `start1 < end2` and `start2 < end1`
3. **Back-to-back**: Meetings ending and starting at same time use same room (use `<=` comparison)
