## Meeting Rooms II

### Problem Statement
Given an array of meeting time intervals `intervals` where `intervals[i] = [start_i, end_i]`, return the **minimum number of conference rooms** required at same time. 

---

### Examples

**Example 1:**
> **Input:** `intervals = [[0, 30], [5, 10], [15, 20]]`  
> **Output:** `2`

**Example 2:**
> **Input:** `intervals = [[7, 10], [2, 4]]`  
> **Output:** `1`

---



# Spec

Dado o param, intervals, array de array de intervalos, onde cada intervalo tem [start,end]
Nosso array não esta ordenado
Retornar o numero máximo de reuniões ao mesmo tempo

intervals = [[0, 30], [5, 10], [15, 20], [20,25]]

 Vai chegar um momento onde teremos até 3 reuniões ao mesmo tempo



```python

import heapq

class Solution:
    def minMeetingRooms(intervals):
        if not intervals:
            return 0
        
        # Sort meetings by start time
        intervals.sort(key=lambda x: x[0])
        
        # Min-heap to store end times of active meetings
        rooms_heap = []
        
        # Add the first meeting's end time
        heapq.heappush(rooms_heap, intervals[0][1])
        
        for interval in intervals[1:]:
            start, end = interval
            # If the earliest room is free, remove it (reuse the room)
            if start >= rooms_heap[0]:
                heapq.heappop(rooms_heap)
                
            # Add the current meeting's end time to the heap
            heapq.heappush(rooms_heap, end)
            
        return len(rooms_heap)
    
```
