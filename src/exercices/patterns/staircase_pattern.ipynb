# Staircase Pattern Exercise

## Problem Description

Create a function to print a staircase pattern using `#` symbols.

The staircase has the following properties:
- Its base and height are both equal to `n`
- It is drawn using `#` symbols and spaces
- The last line is not preceded by any spaces
- All lines are right-aligned

For example, a staircase of size `n = 4`:

```
   #
  ##
 ###
####
```

## Function Description

Complete the `staircase` function with the following parameter(s):

* `int n`: an integer representing the size of the staircase

## Print

Print a staircase as described above. No value should be returned.

## Note
- The last line is not preceded by spaces
- All lines are right-aligned

## Constraints

- `0 < n â‰¤ 100`

## Sample Input

```
6
```

## Sample Output
```
     #
    ##
   ###
  ####
 #####
######
```

````python
def staircase(n):
    """
    Print a staircase pattern of size n
    
    Args:
        n (int): The size of the staircase
        
    Returns:
        None: This function prints the pattern and doesn't return a value
    """
    lineStr = ''.join(' ' for i in range(n))
    index = -1
    while index + 1 > n*-1:
        lineArray = list(lineStr)
        lineArray[index] = '#'
        lineStr = ''.join(lineArray)
        print(lineStr)
        index = index - 1

# Test your function
n = 4
print(f"Staircase of size {n}:")
staircase(n)

# Try with a different size
n = 6
print(f"\nStaircase of size {n}:")
staircase(n)
````
