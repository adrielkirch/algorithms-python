# Staircase Pattern Exercise

## Problem Description

Create a function to print a pyramid pattern using `#` symbols.

The pyramid has the following properties:
- Its base and height are both equal to `n`
- It is drawn using `#` symbols and spaces


For example, a staircase of size `n = 7`:

```
   #
  ###
 #####
#######
```


````python
def pyramid(n):
    arr = [' ']*n
    start = end = n//2
    while start >= 0 and end < n:
        arr[start] = '#'
        arr[end] = '#'
        print(''.join(arr))
        start -= 1
        end  += 1
    
# Test your function
n = 5
print(f"Staircase of size {n}:")
pyramid(n)

# Try with a different size
n = 7
print(f"\nStaircase of size {n}:")
pyramid(n)
````

````python
def pyramid(n):
    height = n // 2 + 1
    for i in range(height):
        spaces = (n // 2) - i
        symbols = 2 * i + 1
        print(' ' * spaces + '#' * symbols )

# Test your function
n = 5
print(f"Staircase of size {n}:")
pyramid(n)

# Try with a different size
n = 7
print(f"\nStaircase of size {n}:")
pyramid(n)
````
