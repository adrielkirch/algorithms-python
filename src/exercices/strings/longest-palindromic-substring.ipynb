## Planejamento

- Poderiamos fazer com brutal force pegando todas as subsitrings que existem em word com dois for aninhados e para cada substring  analisar se substring == substring[::-1] isso vai dar N^3

- Vamos focar numa abordagem que vai ajustando, expandindo a janela para ser  menos custoso em tempo que a abordagem  anterior  

Exemplo:

input word =  'babad'

Iteração i=0, ('b')
    para impar teremos -> 'b' (true)
    para par teremos -> 'ba' (false)
    max = 1

iteração i=1, ('b')
    para impar teremos -> 'b' (true)
        Expande (0,2), ignoramos o -1, teremos 'bab' (true)  
    para par teremos -> 'ab' (false)
    max = 3

iteração i=2, ('b')
    para impar teremos -> 'b' (true)
        Expande (0,2), ignoramos o -1, teremos 'bab' (true)  
    para par teremos -> 'ab' (false)
    max = 3

Iteração i=3 ('a'):
    impar termos -> 'a'
        expandir teremos 'bad' (False)
    para par termos (3,4) -> 'ad'' (False)

nosso  max é 3 'bab' ou outra resposta valida 









````python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        def expand(left, right):        
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right += 1
            return s[left + 1:right]

        result = ""
        for i in range(len(s)):
            # Palíndromo de tamanho ímpar
            palindromic_odd = expand(i, i)
            # Palíndromo de tamanho par
            palindromic_even = expand(i, i + 1)
            # Escolhe o maior
            if len(palindromic_odd) > len(result):
                result = palindromic_odd
            if len(palindromic_even) > len(result):
                result = palindromic_even
        return result



print(Solution().longestPalindrome('babad'))
print(Solution().longestPalindrome('abcdedcbaba'))

````
