The longest comon suffix is the most letters combined from end to start that we can do in all words if non suffix is common return ''

"flower","power","drawer","sawer" is "wer"

```python

def longest_common_suffix(words):
    suffix = words[0]
    for word in words[1:]:
        while not word.endswith(suffix):
            suffix = suffix[1:]
            if not suffix:
                return ''
    return suffix

words = ["flower","power","drawer","sawer"]
print(longest_common_suffix(words))  # "fl"        
```
