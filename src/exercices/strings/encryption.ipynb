## ðŸ§  HeurÃ­stica
- Transformar a string em uma **grade implÃ­cita**
- Preencher **row-wise**
- Ler **column-wise**
- Concatenar os caracteres para gerar o texto criptografado

Esse tipo de problema Ã© comum em desafios de **strings + indexaÃ§Ã£o matemÃ¡tica**.



````python
import math

def encryption(s):
    s = s.replace(" ", "")
    L = len(s)

    rows = math.floor(math.sqrt(L))
    columns = math.ceil(math.sqrt(L))

    if rows * columns < L:
         rows += 1

    matrix = [['' for _ in range(columns)] for _ in range(rows)]
    char_i = 0
    for i in range (0, rows):
        for j in range (0, columns):
            if char_i < L:
                matrix[i][j] = s[char_i] if j < len(s) else ''
                char_i += 1

    encrypted = ''
    for j in range (0, columns):
        for i in range (0, rows):
            encrypted += matrix[i][j]
        encrypted += ' '
    return encrypted[:-1]
             
encryption('if man was meant to stay on the ground god would have given us roots')

````
