# 📝 Problem: Repeated String

There is a string `s` of lowercase English letters that is repeated infinitely many times.

Given an integer `n`, find and print the **number of letter `'a'`** in the **first `n` characters** of this infinite repeated string.

---

## 📌 Example

If:

```
s = "abcac"
n = 10
```

The infinite string becomes:

```
abcacabcacabcac...
```

The substring we consider is the first 10 characters:

```
abcacabcac
```

This contains **4** occurrences of `'a'`.

---

## 📘 Function Description

Implement the function:

```python
def repeatedString(s: str, n: int) -> int:
    pass
```

### **Parameters**

* `s` — string to repeat
* `n` — number of characters to consider in the infinite string

### **Returns**

* `int` — the frequency of `'a'` in the first `n` characters

---

## 📥 Input Format

```
A single string s  
An integer n  
```

---

## 🔒 Constraints

* (1 \le |s| \le 100)
* (1 \le n \le 10^{12})

---

## 🧪 Sample Input 0

```
aba
10
```

### Sample Output 0

```
7
```

### Explanation

The first 10 characters of the infinite string:

```
abaabaabaa
```

There are **7** occurrences of `'a'`.

---

## 🧪 Sample Input 1

```
a
1000000000000
```

### Sample Output 1

```
1000000000000
```

### Explanation

All characters are `'a'`, so the answer is `n`.

````python
def repeatedString(s: str, n: int) -> int:
    if len(s) == 0:
        return 0
    total_a = 0
    for letter in s:
        if letter == 'a':
            total_a += 1
    
    full_repeats = n // len(s)
    total_rest = n % len(s)

    total_a_remaning = 0
    for letter in s[:total_rest]:
        if letter == 'a': 
            total_a_remaning += 1
    return full_repeats * total_a + total_a_remaning

def repeatedStringPythonic(s: str, n: int) -> int:
    if len(s) == 0:
        return 0
    # quantidade de 'a' em uma repetição completa de s
    total_a_in_s = s.count('a')

    # quantas vezes s cabe inteiro dentro dos n caracteres
    full_repeats = n // len(s)

    # quantos caracteres sobram
    remainder = n % len(s)

    # quantidade de 'a' na parte restante
    total_a_remainder = s[:remainder].count('a')

    # total final
    return full_repeats * total_a_in_s + total_a_remainder

# Basic Tests
print(repeatedString("aba", 10))  # Expected: 7
print(repeatedString("a", 1000000000000))  # Expected: 1000000000000
print(repeatedString("abc", 3))  # Expected: 1
print(repeatedString("abc", 5))  # Expected: 1

# Edge Cases
print(repeatedStringPythonic("", 10))  # Empty string (undefined, depends how you handle)
print(repeatedStringPythonic("bcd", 50))  # Expected: 0 (no 'a' characters)
print(repeatedStringPythonic("a", 1))  # Expected: 1
print(repeatedStringPythonic("ba", 1))  # Expected: 0
````
