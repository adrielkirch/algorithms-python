102. Binary Tree Level Order Traversal

Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).




Input: root = [3,9,20,null,null,15,7]
Output: [[3],[9,20],[15,7]]
Example 2:

Input: root = [1]
Output: [[1]]
Example 3:

Input: root = []
Output: []

Constraints:

The number of nodes in the tree is in the range [0, 2000].
-1000 <= Node.val <= 1000

```
        1
      /   \
     2     3
    / \     \
   4   5     6
  / \       /
 7   8     9
```
O segredo desse algoritmo é usar uma Fila (Queue). A lógica é: "Visite o nó, coloque os filhos dele na fila".

Nível 0: Começamos com a raiz [1].

Resultado: [[1]]

Fila para o próximo nível: [2, 3]

Nível 1: Processamos todos que estão na fila (2 e 3).

Resultado: [[1], [2, 3]]

Fila para o próximo nível: [4, 5, 6] (Filhos de 2, depois filhos de 3)

Nível 2: Processamos 4, 5 e 6.

Resultado: [[1], [2, 3], [4, 5, 6]]

Fila para o próximo nível: [7, 8, 9] (Filhos de 4: 7,8; 5 não tem filhos; Filhos de 6: 9)

Nível 3: Processamos 7, 8 e 9.

Resultado Final: [[1], [2, 3], [4, 5, 6], [7, 8, 9]]

````python
from typing import List, Optional
from collections import deque
class Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def queueLevelOrder(self,root: Optional[Node])-> List[List[int]]:
        if root is None:
            return [[]]
        queue = deque([root])
        result = []
        while queue:
            levels = len(queue)
            level_result = []
            for _ in range(levels):
                node = queue.popleft()
                level_result.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            result.append(level_result)
        return result
        

````

````python
# Exemplo de uso (client code)
def build_example_tree():
    # Construindo a árvore:
    #         1
    #       /   \
    #      2     3
    #     / \     \
    #    4   5     6
    #   / \       /
    #  7   8     9
    n7 = Node(7)
    n8 = Node(8)
    n9 = Node(9)
    n4 = Node(4, n7, n8)
    n5 = Node(5)
    n6 = Node(6, n9)
    n2 = Node(2, n4, n5)
    n3 = Node(3, None, n6)
    root = Node(1, n2, n3)
    return root

root = build_example_tree()
sol = Solution()
result = sol.queueLevelOrder(root)
print("Level order traversal:", result)
````
