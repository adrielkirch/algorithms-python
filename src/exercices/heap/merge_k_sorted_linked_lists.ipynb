

# 23. Merge k Sorted Lists

You are given an array of **k** linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

Example 1:

Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted linked list:
1->1->2->3->4->4->5->6
Example 2:

Input: lists = []
Output: []
Example 3:

Input: lists = [[]]
Output: []

```python
from typing import List, Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        values = []
        for list in lists:
            current = list
            while current:
                values.append(current.val)
                current = current.next
        if not values:
            return None
        values.sort()
        head = ListNode(values[0])
        current = head
        for val in values[1:]:
            current.next = ListNode(val, None)
            current = current.next
        return head
```

```python
from typing import List, Optional
import heapq

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        heap = []
        # Adiciona o primeiro nó de cada lista ao heap
        for i, node in enumerate(lists):
            if node:
                # (valor do nó, índice da lista, nó)
                heapq.heappush(heap, (node.val, i, node))
        
        # Nó fictício para facilitar a construção da lista resultante
        result = ListNode()
        current = result
        
        # Enquanto houver elementos no heap
        while heap:
            # Remove o menor elemento do heap
            _, i, node = heapq.heappop(heap)
            # Adiciona esse nó ao resultado
            current.next = node
            # O nó atual agora será o proximo
            current = node
            # Se o nó removido tiver um próximo, adiciona o próximo ao heap
            if node.next:
                heapq.heappush(heap, (node.next.val, i, node.next))
        
        # Retorna o início da lista ligada resultante (ignorando o nó fictício)
        return result.next
```
