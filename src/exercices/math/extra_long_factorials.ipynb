```python
def extraLongFactorials(n):
    """
    Calculate and print the factorial of a given integer.
    
    Args:
        n: An integer for which to calculate the factorial
    
    Returns:
        None (prints the result)
    
    Strategy:
        1. Initialize result to 1
        2. Multiply result by each number from 2 to n
        3. Python handles big integers automatically
        4. Print the final result
    """
    result = 1
    for i in range(2,n+1):
        result *= i
    return result

# Test Case 1: Sample from problem
print("Test Case 1:")
print("Input: 25")
print("Output:")
print(extraLongFactorials(25))
print("Expected: 15511210043330985984000000")
print()

# Test Case 2: Small number
print("Test Case 2:")
print("Input: 5")
print("Output:")
print(extraLongFactorials(5))
print("Expected: 120")
print()

# Test Case 3: Edge case - minimum value
print("Test Case 3:")
print("Input: 1")
print("Output:")
print(extraLongFactorials(1))
print("Expected: 1")
print()

# Test Case 4: Large number
print("Test Case 4:")
print("Input: 30")
print("Output:")
print(extraLongFactorials(30))
print("Expected: 265252859812191058636308480000000")
```
