A **modified Kaprekar number** is a positive whole number with a special property:

* If you square it,
* Then split the resulting number into two parts,
* And sum those parts,
* You obtain the original number.

---

## **Definition**

Consider a positive whole number **n** with **d** digits.

1. Compute **nÂ²**.
2. Let the square have either **2d digits** or **2dâˆ’1 digits**.
3. Split the square into two parts:

   * **r** â†’ the rightmost **d** digits
   * **l** â†’ the remaining left part (may be empty)
4. Convert **l** and **r** to integers.
5. If **l + r == n**, then **n is a modified Kaprekar number**.

> Note: r may contain leading zeros.

---

## **Example**

nÂ² = 81

Split into:

* l = 8

````python
def kaprekarNumbers(p, q):
    found = False
    for num in range(p,q+1):
        d = len(str(num))
        sqr = str(num ** 2)
        left = int(sqr[:-d] or 0)
        right = int(sqr[-d:])

        if left + right == num:
            print(num, end=" ")
            found = True
    if not found:
        print('INVALID RANGE')

kaprekarNumbers(1,99)
        

        




````
