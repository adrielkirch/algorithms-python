# üìò **Taum and B'day ‚Äî Problem Statement (Markdown Completo)**

## üìù Description

Taum is planning to celebrate the birthday of his friend, Diksha. There are two types of gifts she wants:

* **Black gifts**
* **White gifts**

Taum must buy:

* **b** black gifts
* **w** white gifts

Gift prices:

* The cost of each **black** gift is **bc** units
* The cost of each **white** gift is **wc** units
* Converting a gift from black ‚Üí white or white ‚Üí black costs **z** units

Your task is to determine the **minimum total cost** for Taum to obtain all gifts.

---

## üß† Key Insight

Sometimes it's cheaper to:

* Buy all gifts in one color
* Convert some of them to the other color using cost **z**

We compare:

* Buying directly
* Buying the other color + converting

---

## üì• Function Description

Implement:

```python
taumBday(b, w, bc, wc, z)
```

### **Parameters**

| Param | Type | Description                    |

| w     | int  | number of white gifts          |
| bc    | int  | cost of black gift             |
| wc    | int  | cost of white gift             |
| z     | int  | cost to convert between colors |

### **Returns**

`long int`: minimum total cost.

## üß© Explanation Example (Case 3)

```

bc = 9, wc = 1, z = 1
```

* White gifts cost 1 ‚Üí buy directly
* Black gifts cost 9, but:

  * Buy white (1) + convert (1) = **2**, cheaper
* Total = `6*1 + 3*2 = 12`

---

````python
import math
import os
import random
import re
import sys

#
# Complete the 'taumBday' function below.
#
# The function is expected to return a LONG_INTEGER.
# The function accepts following parameters:
#  1. INTEGER b
#  2. INTEGER w
#  3. INTEGER bc
#  4. INTEGER wc
#  5. INTEGER z
#

def taumBday(b, w, bc, wc, z):
    cost_black = min(bc, wc + z)
    cost_white = min(wc, bc + z)
    return b * cost_black + w * cost_white


print(taumBday(10, 10, 1, 1, 1))   # esperado 20
print(taumBday(5, 9, 2, 3, 4))     # esperado 37
print(taumBday(3, 6, 9, 1, 1))     # esperado 12
print(taumBday(7, 7, 4, 2, 1))     # esperado 35
print(taumBday(3, 3, 1, 9, 2))     # esperado 12
````
