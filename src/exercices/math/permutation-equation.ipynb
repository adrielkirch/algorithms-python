# Sequence Equation

## Problem Statement

You are given an array **p** of **n** distinct integers where each element is between 1 and n.

**Your task**: For each number **x** from 1 to n, find a number **y** such that **p[p[y]] = x**.

Think of **p** as a function where **p(i)** gives you the value at position **i**.

## What does p[p[y]] mean?
- **p[y]** = the value at position **y** in array **p**
- **p[p[y]]** = use that value as an index to get another value from **p**

## Simple Example

Let's say **p = [2, 3, 1]** (positions 1, 2, 3 contain values 2, 3, 1)

**Goal**: Find **y** for each **x** such that **p[p[y]] = x**

**For x = 1**: We need **p[p[y]] = 1**
- Try y = 2: p[2] = 3, then p[3] = 1 âœ“ 
- So when x = 1, answer is y = 2

**For x = 2**: We need **p[p[y]] = 2**  
- Try y = 3: p[3] = 1, then p[1] = 2 âœ“
- So when x = 2, answer is y = 3

**For x = 3**: We need **p[p[y]] = 3**
- Try y = 1: p[1] = 2, then p[2] = 3 âœ“  
- So when x = 3, answer is y = 1

**Answer**: [2, 3, 1]

## Another Way to Think About It

1. **Question**: "What position gives me x when I do the double lookup p[p[y]]?"
2. **Process**: 
   - Find where x appears in the array â†’ this tells us what p[y] should be
   - Find where that value appears â†’ this tells us what y should be

# ðŸ“˜ Understanding the Heuristic Behind **Permutation Equation**

### *A step-by-step guide to turning a confusing problem into a simple algorithm*

---

## ðŸ” Problem Summary

Given a permutation `p` of integers from `1` to `n`, for each value `x` from `1` to `n`, you must find a value `y` such that:

```
p[p[y]] = x
```

The result is an array of all these `y` values.

---

# ðŸ§  Why This Problem Confuses Many People

Even though the input is small and the operations seem simple, the logic is **not obvious**:

* You must work with **two levels of indexing** (`p[p[y]]`).
* Input is **1-based**, Python lists are **0-based**.
* The key step â€” *building the inverse permutation* â€” is not something beginners do naturally.

This guide breaks it down into a practical heuristic.

---

# ðŸŽ¯ Heuristic Overview

### âœ”ï¸ The problem boils down to:

> **Given a permutation, find where each number appears.
> Then apply the mapping twice.**

To do this efficiently, you build an array:

```
pos[value] = index where that value occurs
```

This is the **inverse permutation**.

---

# ðŸ§© Step 1 â€” Build the `pos[]` Array

### Why?

We want to quickly answer:

> â€œWhere is the value `v` inside array `p`?â€

Since `p` is a permutation, each value is unique, so its position is well-defined.

### How?

```python
pos = [0] * (n + 1)

for i in range(n):
    pos[p[i]] = i + 1   # +1 because we want 1-based positions
```

### Example

Given:

```
p = [5, 2, 1, 3, 4]
```

We build:

```
value: 1 2 3 4 5
pos:   3 2 4 5 1
```

This means:

* 1 is in position 3
* 2 is in position 2
* 5 is in position 1
* etc.

---

# ðŸ§© Step 2 â€” Apply the Permutation Twice

The equation:

```
p[p[y]] = x
```

Rewritten using `pos[]`:

1. Find `i` such that `p[i] = x`
   â†’ `i = pos[x]`

2. Find `y` such that `p[y] = i`
   â†’ `y = pos[i]`

Final answer for each `x` is:

```
y = pos[pos[x]]
```

---

# ðŸ§® Step 3 â€” Generate Results

```python
result = []

for x in range(1, n+1):
    first = pos[x]      # where x occurs
    second = pos[first] # where that index occurs
    result.append(second)
```

# ðŸ§ª Complete Working Solution

```python
def permutationEquation(p):
    n = len(p)
    pos = [0] * (n + 1)

    # Build inverse permutation
    for i in range(n):
        pos[p[i]] = i + 1

    # Compute final answer
    result = []
    for x in range(1, n + 1):
        i = pos[x]
        y = pos[i]
        result.append(y)
    return result
```

> **Build an inverse permutation array `pos[value] = index`,
> then for each `x`, compute `pos[pos[x]]`.**

Once you internalize this pattern, problems like this become simple and elegant.
