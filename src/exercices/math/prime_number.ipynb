 # Prime number
 
 **Given a any number > 0 it should return True if its a prime else False**
 
 A number is **prime** if its only **divided by itself** and  **1**.

 **1** is not considered prime

 **2** is the unique Odd number considered a prime
 
 **3** is prime

 **5** is prime

 **7** is prime

 **9** is not a prime is has more than 2 divisors

### Exericise

 Find all prime numbers and show the difference between current_prime and previous prime to check the intervals between them.

```python

def isPrime(num: int) -> bool:

    if num == 1:
        return False
    elif num == 2:
        return True
    elif num % 2 == 0:
         return False
    
    divisor = num
    count_divisors = 0

    while divisor >= 1:
        if num % divisor == 0:
            count_divisors = count_divisors + 1
        divisor = divisor - 2
    
    if count_divisors == 2:
        return True
    return False
        
prime_list=[]
for i in range(1, 10000):
    if isPrime(i):
        prime_list.append(i)
        if len(prime_list) > 1:
            dif = prime_list[-1] - prime_list[-2]
            print(f"{prime_list[-1]} - Difference between previous prime: {dif}")
        else:
            print(f"{prime_list[-1]}")

print("Total found ->",len(prime_list))

```
