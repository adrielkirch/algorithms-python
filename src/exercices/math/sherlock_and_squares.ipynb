`````python
import math

#
# Complete the 'squares' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER a
#  2. INTEGER b
#

def squares(a, b):
    count = 0
    for num in range(a,b+1):
        root = math.sqrt(num)
        if root.is_integer():
            count += 1
    return count


`````

`````python
import math

def squares(a, b):
    count = 0
    for num in range(a,b+1):
        if math.isqrt(num):
            count += 1
    return count
`````

`````python

import math

def squares(a, b):
    print(f"{math.sqrt(a)} and {math.sqrt(b)}")
    min_root = math.ceil(math.sqrt(a)) # ceil to round up
    max_root = math.floor(math.sqrt(b)) # floor to round down
    total = max_root - min_root + 1
    return total if total > 0 else 0
`````
