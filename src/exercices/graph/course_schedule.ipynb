## 📝 Enunciado

Existem um total de `numCourses` cursos que você deve cursar, rotulados de `0` a `numCourses - 1`. Você recebe um array `prerequisites` onde `prerequisites[i] = [a, b]` indica que você **deve** fazer o curso `b` primeiro se quiser fazer o curso `a`.

* Por exemplo, o par `[0, 1]` indica que, para fazer o curso `0`, você precisa ter concluído o curso `1`.

Retorne **`True`** se for possível terminar todos os cursos. Caso contrário, retorne **`False`**.

---

## 💡 Exemplos

### **Exemplo 1:**
* **Input:** `numCourses = 2`, `prerequisites = [[1, 0]]`
* **Output:** `true`
* **Explicação:** Existem 2 cursos no total. Para fazer o curso 1, você deve ter terminado o curso 0. Isso é possível.

### **Exemplo 2:**
* **Input:** `numCourses = 2`, `prerequisites = [[1, 0], [0, 1]]`
* **Output:** `false`
* **Explicação:** Existem 2 cursos no total. Para fazer o curso 1, você deve ter terminado o curso 0, e para fazer o curso 0, você deve ter terminado o curso 1. Isso cria um **ciclo impossível**.

---

## ⚙️ Restrições
* `1 <= numCourses <= 2000`
* `0 <= prerequisites.length <= 5000`
* `prerequisites[i].length == 2`
* `0 <= ai, bi < numCourses`
* Todos os pares `prerequisites[i]` são **únicos**.

---



## 🧠 Heurística de Resolução (Uber Level)

1.  **Representação:** O problema é um **Grafo Direcionado**. Cada curso é um nó e cada pré-requisito é uma aresta (seta) indo do pré-requisito para o curso dependente.
2.  **O Problema Real:** O exercício pede para detectar se existe um **CICLO** no grafo. Se houver ciclo, não há ordem lógica possível.
3.  **Técnicas Comuns:**
    * **DFS (Depth First Search):** Usando estados (Visitando/Visitado) para detectar se "batemos" em um nó que ainda está na pilha de recursão.
    * **Algoritmo de Kahn (BFS):** Usando o conceito de *Indegree* (grau de entrada). Cursos sem pré-requisitos começam com indegree 0.

````python
from typing import List
from collections import defaultdict

UNVISITED= 0
VISITING = 1
FINISHED= 2

class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        # Usando defaultdict para lista de adjacência
        adj = defaultdict(list)

        for course, prerequisite in prerequisites:
            adj[prerequisite].append(course)

        # Estados: 0 = não visitado, 1 = visitando, 2 = visitado
        state = [UNVISITED] * numCourses

        def dfs(node):
            if state[node] == VISITING:
                # Encontrou ciclo
                return False
            elif state[node] == FINISHED:
                # Já processado, sem ciclo
                return True
            state[node] = VISITING  # Marcando como visitando
            for neighbor in adj[node]:
                valid_graph = dfs(neighbor)
                if not valid_graph:
                    return False
            state[node] = FINISHED
            return True

        for i in range(numCourses):
            if not dfs(i):
                return False
        return True
````

````python
# Testes para a função canFinish
def test_canFinish():
    sol = Solution()
    # Caso 1: Possível terminar todos os cursos
    numCourses1 = 2
    prerequisites1 = [[1, 0]]
    print(f"Teste 1 - Esperado: True | Resultado: {sol.canFinish(numCourses1, prerequisites1)}")

    # Caso 2: Ciclo impossível
    numCourses2 = 2
    prerequisites2 = [[1, 0], [0, 1]]
    print(f"Teste 2 - Esperado: False | Resultado: {sol.canFinish(numCourses2, prerequisites2)}")

    # Caso 3: Nenhum pré-requisito
    numCourses3 = 3
    prerequisites3 = []
    print(f"Teste 3 - Esperado: True | Resultado: {sol.canFinish(numCourses3, prerequisites3)}")

    # Caso 4: Grafo maior sem ciclo
    numCourses4 = 4
    prerequisites4 = [[1, 0], [2, 1], [3, 2]]
    print(f"Teste 4 - Esperado: True | Resultado: {sol.canFinish(numCourses4, prerequisites4)}")

    # Caso 5: Grafo maior com ciclo
    numCourses5 = 4
    prerequisites5 = [[1, 0], [2, 1], [0, 2], [3, 2]]
    print(f"Teste 5 - Esperado: False | Resultado: {sol.canFinish(numCourses5, prerequisites5)}")

test_canFinish()
````
