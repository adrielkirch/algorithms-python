There are a total of numCourses courses you have to take, labeled from 0 to numCourses - 1. You are given an array prerequisites where prerequisites[i] = [ai, bi] indicates that you must take course bi first if you want to take course ai.

For example, the pair [0, 1], indicates that to take course 0 you have to first take course 1.
Return the ordering of courses you should take to finish all courses. If there are many valid answers, return any of them. If it is impossible to finish all courses, return an empty array.

Example 1:

Input: numCourses = 2, prerequisites = [[1,0]]
Output: [0,1]
Explanation: There are a total of 2 courses to take. To take course 1 you should have finished course 0. So the correct course order is [0,1].
Example 2:

Input: numCourses = 4, prerequisites = [[1,0],[2,0],[3,1],[3,2]]
Output: [0,2,1,3]
Explanation: There are a total of 4 courses to take. To take course 3 you should have finished both courses 1 and 2. Both courses 1 and 2 should be taken after you finished course 0.
So one correct course order is [0,1,2,3]. Another correct ordering is [0,2,1,3].
Example 3:

Input: numCourses = 1, prerequisites = []
Output: [0]

# Topological sort


```python
# Importa defaultdict para criar listas de adjacência facilmente
from collections import defaultdict

# Constantes para marcar o estado de cada nó durante a busca
UNVISITED = 0  # Nunca visitado
VISITING = 1   # Em visitação (na pilha de recursão)
VISITED = 2    # Já totalmente processado

def findOrder(numCourses, prerequisites):
    """
    Retorna uma ordem possível para completar todos os cursos,
    respeitando os pré-requisitos. Se não for possível (ciclo), retorna [].
    """
    # Cria lista de adjacência: cada curso aponta para os cursos que dependem dele
    # Exemplo: se prerequisites = [[1,0]], então 0 -> 1 (para fazer 1, precisa de 0 antes)
    adjacency = defaultdict(list)
    for course, prerequisite in prerequisites:
        adjacency[prerequisite].append(course)

    result = []  # Lista para armazenar a ordem dos cursos
    states = [UNVISITED] * numCourses  # Estado de cada curso

    def dfs(index):
        """
        Retorna True se não há ciclo a partir deste nó, False se ciclo detectado.
        Marca o estado do nó durante a busca para detectar ciclos.
        """
        if states[index] == VISITING:
            # Se já estamos visitando este nó, há um ciclo no grafo
            return False # Ciclo detectado
        elif states[index] == VISITED:
            # Se já visitamos totalmente este nó, não precisa processar de novo
            return True

        # Marca como em visitação (na pilha de recursão)
        states[index] = VISITING

        # Visita todos os vizinhos (cursos que dependem deste)
        for neighbor in adjacency[index]:
            if not dfs(neighbor):
                # Se algum vizinho formar ciclo, propaga o erro para cima
                return False

        # Marca como totalmente visitado (sai da pilha de recursão)
        states[index] = VISITED
        result.append(index)
        return True

    # Tenta visitar todos os cursos (pode haver componentes desconexos)
    for i in range(numCourses):
        if states[i] == UNVISITED:
            if not dfs(i):
                return []


    return result  

# Exemplo de teste: deve retornar algo como [0, 2, 1, 3] ou [0, 1, 2, 3]
print(findOrder(4, [[1,0],[2,0],[3,1],[3,2]]))  # Teste didático
```
