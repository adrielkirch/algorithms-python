## 🧩 Number of Connected Components in an Undirected Graph


You are given an integer n representing the number of nodes in an undirected graph. The nodes are labeled from **0** to **n - 1**.

You are also given an array edges where **edges[i] = [ai, bi]** indicates that there is an undirected edge between node ai and node bi.

Return the number of connected components in the graph.

A connected component is a group of nodes such that there is a path between any two nodes in the group, and no node in the group is connected to any node outside the group.

🧪 Example 1:

Input: n = 5, edges = [[0,1],[1,2],[3,4]]
Output: 2
Explanation:
Nodes 0, 1, and 2 are connected → 1 component
Nodes 3 and 4 are connected → 1 component
Total = 2 components

🧪 Example 2:

Input: n = 5, edges = [[0,1],[1,2],[2,3],[3,4]]
Output: 1
Explanation:
All nodes are connected → 1 component

# 1️⃣ Specification (Especificação de Engenharia)

**Core do Problema:**
Dado um grafo não direcionado com `n` nós (rotulados de 0 a n-1) e uma lista de arestas, determine quantos componentes conectados existem, ou seja, quantos grupos de nós estão interligados entre si, mas não com outros grupos.

**Edge Cases & Pegadinhas:**
1. Grafo totalmente desconexo (nenhuma aresta): cada nó é um componente.
2. Grafo totalmente conectado (todas as arestas possíveis): apenas 1 componente.
3. Grafo com laços (arestas do tipo [a, a]): não afeta a contagem de componentes.
4. Grafo com múltiplas arestas repetidas: não deve contar componentes extras.
5. Grafo com nós isolados (nós sem nenhuma aresta): cada um é um componente separado.

# 2️⃣ Plan (A Heurística e o Desenho Técnico)

**Brute Force (Intuição):**
Poderíamos tentar, para cada nó, buscar todos os outros nós conectados a ele, marcando-os como visitados. Mas, se fizermos isso sem controle, podemos visitar os mesmos nós várias vezes, tornando a solução ineficiente (O(N^2) ou pior).

**Heurística Otimizada:**
Vamos usar **DFS (Depth-First Search)** para "explorar" cada componente. A cada vez que encontramos um nó não visitado, iniciamos uma nova busca (DFS), marcando todos os nós desse componente. O número de vezes que iniciamos uma nova DFS é igual ao número de componentes conectados.

**Por que DFS vence?**
DFS (ou BFS) garante que cada nó seja visitado apenas uma vez, e cada aresta também, resultando em tempo O(V + E), ideal para grafos esparsos e densos.

# 4️⃣ Complexity (Veredito Final)

- **Time Complexity:** O(V + E), pois cada nó e cada aresta é visitado no máximo uma vez.
- **Space Complexity:** O(V + E) para armazenar o grafo e o conjunto de visitados (stack pode chegar a O(V) em grafos "esticados").

---
Dúvidas? Teste com grafos desconexos, laços e arestas duplicadas para garantir robustez!

````python
from typing import List

class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:
        # Construir o grafo como lista de adjacência
        graph = {i: [] for i in range(n)}
        for a, b in edges:
            graph[a].append(b)
            graph[b].append(a)
        
        visited = set()
        total = 0
        
        def dfs(node: int) -> None:
            stack = [node]
            while stack:
                curr = stack.pop()
                if curr in visited:
                    continue
                visited.add(curr)
                neighbors = graph[curr]
                # Adiciona todos os vizinhos não visitados
                for neighbor in neighbors:
                    if neighbor not in visited:
                        stack.append(neighbor)
            # O(V + E) para cada componente
        
        for i in range(n):
            if i in visited:
                continue
            dfs(i)
            total += 1  # Cada chamada nova é um componente
        return total  # O(V + E) no total
````
