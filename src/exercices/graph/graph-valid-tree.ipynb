## 261. Graph Valid Tree

Medium | Topics: Graph, Depth-First Search, Breadth-First Search, Union Find

You have a graph of n nodes labeled from 0 to n - 1. You are given an integer n and a list of edges where edges[i] = [ai, bi] indicates that there is an undirected edge between nodes ai and bi in the graph.

Return true if the edges of the given graph make up a valid tree, and false otherwise.

###   Example 1:

Input: n = 5, edges = [[0,1],[0,2],[0,3],[1,4]]
Output: true

### Example 2:

Input: n = 5, edges = [[0,1],[1,2],[2,3],[1,3],[1,4]]
Output: false

Constraints:

1 <= n <= 2000

0 <= edges.length <= 5000

edges[i].length == 2

0 <= ai, bi < n

ai != bi

There are no self-loops or repeated edge

#   Specification & Heurística

## Intro

- Verificar que um grafo não direcionado é uma árvore
- Não pode ter um ciclo
- Uma árvore com N nós tem exatamente N-1 arestas

## Edge cases

- Um nó isolado sozinho é uma tree se não apontar a si mesmo
- Grafo desconectado não é uma arvore, mesmo sem ciclo
- Caso o numero de arestas for maior que n-1 então há um ciclo

## Heurística de Solução

Uma árvore válida DEVE satisfazer 3 condições:

### 1️⃣ Número correto de arestas
- Uma árvore com `n` nós tem **exatamente** `n-1` arestas
- Se `len(edges) ≠ n-1` → **não é árvore**
```
n=5, edges=4 ✓ (pode ser árvore)
n=5, edges=3 ✗ (desconectado)
n=5, edges=5 ✗ (tem ciclo)
```

### 2️⃣ Sem ciclos (detectar com DFS)
- Usar DFS rastreando nó **pai**
- Se encontrar vizinho visitado que **não é o pai** → há ciclo
- Razão: Em grafo não direcionado, a aresta de volta ao pai é bidirecional, não é um ciclo

```
Exemplo: 0 -- 1
adjacency[0] = [1]
adjacency[1] = [0]

DFS(0, parent=-1):
  ├─ vizita 1
  └─ DFS(1, parent=0):
     ├─ encontra vizinho 0
     └─ 0 está visitado E é o pai → OK!
```

### 3️⃣ Grafo conectado
- Todos os `n` nós devem ser visitados
- Se `len(visited) ≠ n` → há componentes desconectadas

```
n=5, visita 3 nós ✗ (2 nós isolados)
n=5, visita 5 nós ✓ (conectado)
```

## Plan

### Step 1: Verificação de arestas
- `if n - 1 != len(edges): return False`

### Step 2: Construir adjacency list
```python
adjacency_list[u].append(v)
adjacency_list[v].append(u)  # bidirectional
```

### Step 3: DFS com rastreamento de pai
```python
def dfs(node, parent):
    for neighbor in adjacency_list[node]:
        if neighbor == parent:
            continue  # ignore aresta de volta
        if neighbor in visited:
            return False  # ciclo detectado
        dfs(neighbor, node)
```

### Step 4: Verificar conectividade
- `return len(visited) == n`


````python
## Abordagem MAIS SIMPLES: DFS Minimalista
from Typing import List
class Solution_Simple:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        # Condição 1: Arestas corretas
        if len(edges) != n - 1:
            return False
        
        # Construir grafo
        graph = [[] for _ in range(n)]
        for u, v in edges:
            graph[u].append(v)
            graph[v].append(u)
        
        # Condição 2 e 3: DFS para verificar ciclos + conectividade
        visited = set()
        
        def dfs(node, parent):
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor == parent:  # Ignorar aresta de volta
                    parent = -1  # Marca como processado
                    continue
                if neighbor in visited:
                    return False  # Ciclo!
                if not dfs(neighbor, node):
                    return False
            return True
        
        # Começar do nó 0 - se houver ciclo, retorna False
        if not dfs(0, -1):
            return False
        
        # Verificar conectividade
        return len(visited) == n

# Teste
print("=== ABORDAGEM SIMPLES ===")
sol_simple = Solution_Simple()

print("\nTest 1: Árvore válida")
result = sol_simple.validTree(5, [[0,1], [0,2], [0,3], [1,4]])
print(f"  Resultado: {result} (esperado: True) ✓" if result else f"  Resultado: {result} ✗")

print("\nTest 2: Com ciclo")
result = sol_simple.validTree(5, [[0,1], [1,2], [2,3], [1,3], [1,4]])
print(f"  Resultado: {result} (esperado: False) ✓" if not result else f"  Resultado: {result} ✗")

print("\nTest 3: Desconectado")
result = sol_simple.validTree(4, [[0,1], [1,2]])
print(f"  Resultado: {result} (esperado: False) ✓" if not result else f"  Resultado: {result} ✗")
````
