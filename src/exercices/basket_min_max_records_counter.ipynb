# 🏀 Breaking Maria's Basketball Records

Maria plays college basketball and wants to go pro. Each season she tracks her **highest** and **lowest** points scored per game. She counts the number of times she breaks her season records.

The **first game** sets the initial record for both highest and lowest points.

---

## Example Scores

```python
scores = [12, 24, 10, 24]
```

Let's visualize her record progress:

| Game | Score | Min Record | Max Record | Min Broken | Max Broken |
| ---- | ----- | ---------- | ---------- | ---------- | ---------- |
| 0    | 12    | 12         | 12         | 0          | 0          |
| 1    | 24    | 12         | 24         | 0          | 1 ✅       |
| 2    | 10    | 10         | 24         | 1 ✅       | 1          |
| 3    | 24    | 10         | 24         | 1          | 1          |

**Legend:**
- ✅ indicates the record was broken.

---

## Problem

Given an array of scores, determine:

1. How many times Maria breaks her **most points** record.
2. How many times she breaks her **least points** record.

---

## Function Signature

```python
def breakingRecords(scores):
    """
    Returns a list [max_breaks, min_breaks] indicating the number of times
    the max and min records were broken.
    """
```

### Parameters

- `scores[n]`: list of integers representing points per game.

### Returns

- List of two integers: `[max_breaks, min_breaks]`

---

## Example

```python
scores = [12, 24, 10, 24]
result = breakingRecords(scores)
print(result)  # Output: [1, 1]
```

**Explanation:**

- **Max record broken:** Game 1 (24 > 12) → 1 time
- **Min record broken:** Game 2 (10 < 12) → 1 time

---

## Input Format

1. Integer `n` → number of games
2. Array of `n` integers → scores per game

```
4
12 24 10 24
```

---

## Constraints

- `1 ≤ n ≤ 1000`
- `0 ≤ scores[i] ≤ 10^8`

---

## Step-by-Step Approach

1. Initialize `min_score` and `max_score` with the first game's score.
2. Initialize counters `min_count` and `max_count` to `0`.
3. Iterate over the scores starting from the second game:
   - If current score > `max_score`, update `max_score` and increment `max_count`.
   - If current score < `min_score`, update `min_score` and increment `min_count`.
4. Return `[max_count, min_count]`.
