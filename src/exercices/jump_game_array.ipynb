You are given an integer array nums. You are initially positioned at the array's first index, and each element in the array represents your maximum jump length at that position.

Return true if you can reach the last index, or false otherwise.
s

Example 1:

Input: nums = [2,3,1,1,4]
Output: true
Explanation: Jump 1 step from index 0 to 1, then 3 steps to the last index.

Example 2:

Input: nums = [3,2,1,0,4]
Output: false
Explanation: You will always arrive at index 3 no matter what. Its maximum jump length is 0, which makes it impossible to reach the last index.

 

```python

def canJump(nums):
    max_reach = 0
    n = len(nums)
    for i in range(n):
        if i > max_reach:
            return False  
        max_reach = max(max_reach,nums[i] + i)
    return True

nums = [3,2,1,0,4]
print(canJump(nums))
```

You are given an integer array coins representing coins of different denominations and an integer amount representing a total amount of money.

Return the fewest number of coins that you need to make up that amount. If that amount of money cannot be made up by any combination of the coins, return -1.

You may assume that you have an infinite number of each kind of coin.

 

Example 1:

Input: coins = [1,2,5], amount = 11
Output: 3
Explanation: 11 = 5 + 5 + 1

Example 2:

Input: coins = [2], amount = 3
Output: -1

Example 3:

Input: coins = [1], amount = 0
Output: 0

 

Constraints:

    1 <= coins.length <= 12
    1 <= coins[i] <= 2^31 - 1
    0 <= amount <= 10^4

```python
Dynamic Programming = solves a large problem using solutions to smaller problems.

Large problem: form quantity = 11 using coins [1, 2, 5].

Subproblem: form any value less than 11, because we can construct 11 from them.

Practical heuristic:

Think of the smallest case (0) → DP basis

For each value from 1 to value, ask:

"If I use each coin, what would be the minimum number of coins needed?"

dp[i] = minimum number of coins to form i.

dp = [0, inf, inf, inf, inf, inf, inf, inf, inf, inf, inf, inf]

dp[0] = 0 because 0 coins forms 0

3️⃣ fill the dp

let's loop 1 to 11

i = 1
coin = 1 → 1 ≤ i → dp[1] = min(dp[1], dp[1-1]+1) = min(inf, 0+1) = 1
coin = 2 → 2 > 1 → ignora
coin = 5 → 5 > 1 → ignora



dp: [0, 1, inf, inf, inf, inf, inf, inf, inf, inf, inf, inf]


i = 2

coin = 1 → dp[2] = min(dp[2], dp[2-1]+1) = min(inf, 1+1) = 2
coin = 2 → dp[2] = min(2, dp[2-2]+1) = min(2, 0+1) = 1

i = 3

coin = 1 → dp[3] = min(infd, dp[3-1]+1) = min(inf, 1+1) = 2
coin = 2 → dp[3] = min(2, dp[3-2]+1) = min(2, 1+1) = 2
coin = 5 → ignora
dp: [0, 1, 1, 2, inf, inf, inf, inf, inf, inf, inf, inf]

i = 5

coin = 1 → dp[5] = min(inf, dp[5-1]+1) = min(inf, 2+1) = 3
coin = 2 → dp[5] = min(3, dp[5-2]+1) = min(3, 2+1) = 3
coin = 5 → dp[5] = min(3, dp[5-5]+1) = min(3, 0+1) = 1

dp: [0, 1, 1, 2, 2, 1, inf, inf, inf, inf, inf, inf]

i = 6: dp[6] = min(dp[5]+1, dp[4]+1, dp[1]+1) = 2
i = 7: dp[7] = min(dp[6]+1, dp[5]+1, dp[2]+1) = 3
i = 8: dp[8] = min(dp[7]+1, dp[6]+1, dp[3]+1) = 3
i = 9: dp[9] = min(dp[8]+1, dp[7]+1, dp[4]+1) = 4
i = 10: dp[10] = min(dp[9]+1, dp[8]+1, dp[5]+1) = 2
i = 11: dp[11] = min(dp[10]+1, dp[9]+1, dp[6]+1) = 3
```

```python
class Solution:
    def coinChange(self, coins, amount: int) -> int:
        dp = [float('inf')] * (amount + 1)
        # Base case: 0 coins are needed to make amount 0
        dp[0] = 0
        for i in range(1, amount+1):
            for coin in coins:
                if i - coin >= 0:
                    dp[i] = min(dp[i], dp[i - coin] + 1)
        return dp[amount] if dp[amount] != float('inf') else -1

Solution().coin()
```
