
# ðŸ“˜ Library Fine â€” Problem Explanation

This is a classic **ad-hoc simulation** problem (also appears as **greedy / implementation** in competitive programming).

You must compute the fine for a library book based only on the *least precise* unit of lateness:

### ðŸ“ Fine Rules

Let **Returned date = d1, m1, y1**
Let **Due date = d2, m2, y2**

---

## ðŸ”¹ 1. Returned **after the expected year**

If **y1 > y2**:

**fine = 1000**
---

## ðŸ”¹ 2. Returned **after the expected month**, but same year

If **y1 == y2** and **m1 > m2**:

**fine = 15 * (d1 - d2)**

---

## ðŸ”¹ 3. Returned **after the expected day**, but same month & year

If **y1 == y2** and **m1 == m2** and **d1 > d2**:

**fine = 15 * (d1 - d2)**

---

## ðŸ”¹ 4. Otherwise

Returned on time:

**fine = 0**

You should use a common technique in conditional logic called:

âœ… "Short-circuiting by funneling"

Also known as:

âœ” Decision Cascade
âœ” Sequential Filtering
âœ” Guard Clauses
âœ” Early Return

ðŸ“Œ What's happening in your code?

You are:

Comparing the most important criterion first **(year).** then **month** then **day**

Quickly eliminating possibilities (guard clauses):

If it's before â†’ return

If it's after â†’ return

The implicit "else" means:

If it didn't fall into the < or >, it's because they are equal.

Only then do you "funnel" to the next level (month).

And then to the most specific level (day).

```python
def libraryFine(d1, m1, y1, d2, m2, y2):
    # Ano: decide tudo
    if y1 < y2:
        return 0
    elif y1 > y2:
        return 10000
    
    # MÃªs (ano igual)
    if m1 < m2:
        return 0
    elif m1 > m2:
        return 500 * (m1 - m2)

    # Dia (ano e mÃªs iguais)
    if d1 <= d2:
        return 0
    
    return 15 * (d1 - d2)

print("Same day return (no fine):", libraryFine(1, 1, 2020, 1, 1, 2020))  # 0
print("Returned earlier:", libraryFine(1, 1, 2020, 5, 1, 2020))  # 0

# Day late (same month/year)
print("1 day late:", libraryFine(10, 1, 2020, 5, 1, 2020))  # 0 (bug here, according to original logic)

# Month late (same year)
print("1 month late:", libraryFine(1, 3, 2020, 1, 2, 2020))  # 500

# Several months late
print("3 months late:", libraryFine(1, 6, 2020, 1, 3, 2020))  # 500 * 3

# Year late
print("1 year late:", libraryFine(1, 1, 2021, 31, 12, 2020))  # 10000

# Earlier year (no fine)
print("Returned previous year:", libraryFine(1, 1, 2019, 5, 1, 2020))  # 0

# Edge cases at boundaries
print("End of year return:", libraryFine(1, 1, 2021, 31, 12, 2020))  # 10000
print("End of month return:", libraryFine(1, 2, 2021, 31, 1, 2021))  # 500
```
