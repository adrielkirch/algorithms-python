This is a classic **ad-hoc simulation** problem (also appears as **greedy / implementation** in competitive programming).

You must compute the fine for a library book based only on the *least precise* unit of lateness:

### ðŸ“ Fine Rules

Let **Returned date = d1, m1, y1**
Let **Due date = d2, m2, y2**

---

## ðŸ”¹ 1. Returned **after the expected year**

If **y1 > y2**:

**fine = 1000**
---

## ðŸ”¹ 2. Returned **after the expected month**, but same year

If **y1 == y2** and **m1 > m2**:

**fine = 15 * (d1 - d2)**

---

## ðŸ”¹ 3. Returned **after the expected day**, but same month & year

If **y1 == y2** and **m1 == m2** and **d1 > d2**:

**fine = 15 * (d1 - d2)**

---

## ðŸ”¹ 4. Otherwise

Returned on time:

**fine = 0**

```python
def libraryFine(d1, m1, y1, d2, m2, y2):
    # Ano: decide tudo
    if y1 < y2:
        return 0
    elif y1 > y2:
        return 10000
    
    # MÃªs (ano igual)
    if m1 < m2:
        return 0
    elif m1 > m2:
        return 500 * (m1 - m2)

    # Dia (ano e mÃªs iguais)
    if d1 > d2:
        return 15 * (d1 - d2)
    
    return 0
```
