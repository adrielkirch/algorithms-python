## 153. Find Minimum in Rotated Sorted Array

Suppose an array of length n sorted in ascending order is rotated between 1 and n times. 

For example, the array nums = [0,1,2,4,5,6,7] might become:

    [4,5,6,7,0,1,2] if it was rotated 4 times.
    [0,1,2,4,5,6,7] if it was rotated 7 times.

    

Notice that rotating an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

[0, 1, 2, 4, 5, 6, 7] 0 rotation
[7, 0, 1, 2, 4, 5, 6] 1 rotation
[6, 7, 0, 1, 2, 4, 5] 2 rotation
[5, 6, 7, 0, 1, 2, 4] 3 rotation
[4, 5, 6, 7, 0, 1, 2] 4 rotation
[2, 4, 5, 6, 7, 0, 1] 5 rotation
[1, 2, 4, 5, 6, 7, 0] 6 rotation
[0, 1, 2, 4, 5, 6, 7] 7 rotation

Given the sorted rotated array nums of unique elements, return the minimum element of this array.

You must write an algorithm that runs in O(log n) time.


**Example 1:**

Input: nums = [3,4,5,1,2]
Output: 1
Explanation: The original array was [1,2,3,4,5] rotated 3 times.

**Example 2:**

Input: nums = [4,5,6,7,0,1,2]
Output: 0
Explanation: The original array was [0,1,2,4,5,6,7] and it was rotated 4 times.

**Example 3:**

Input: nums = [11,13,15,17]
Output: 11
Explanation: The original array was [11,13,15,17] and it was rotated 4 times. 

 
Constraints:

    n == nums.length
    1 <= n <= 5000
    -5000 <= nums[i] <= 5000
    All the integers of nums are unique.
    nums is sorted and rotated between 1 and n times.

````python
class Solution(object):
    def findMin(self, nums):

        left, right = 0, len(nums) - 1

        # Se o array já está ordenado (sem rotação), o menor é o primeiro elemento
        if nums[left] < nums[right]:
            return nums[left]
        
        while left < right:
            mid = (left + right) // 2
            # Se o do meio for maior que o último, o menor está na DIREITA
            if nums[mid] > nums[right]:
                left = mid + 1
            # Caso contrário, o menor está na ESQUERDA (incluindo o mid)
            else:
                right = mid
        return nums[left] # Retorna o VALOR real do menor elemento

sol = Solution()

# Caso 1: Array rotacionado
teste1 = [4,5,6,7,0,1,2]
print(f"Teste 1: {teste1}")
print(f"O que o LeetCode quer: 1")
print(f"O que seu código retorna: {sol.findMin(teste1)}") 
print("-" * 30)
# Caso 2: Array já ordenado
teste2 = [11, 13, 15, 17]
print(f"Teste 2: {teste2}")
print(f"O que o LeetCode quer: 11")
print(f"O que seu código retorna: {sol.findMin(teste2)}")
print("-" * 30)
# Caso 3: Array com a quebra no final
teste3 = [2, 3, 4, 0]
print(f"Teste 3: {teste3}")
print(f"O que o LeetCode quer: 0")
print(f"O que seu código retorna: {sol.findMin(teste3)}")
                
````
