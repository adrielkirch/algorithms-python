# Exercício: Busca Binária em Array Ordenado

Implemente uma função em Python que realize a busca binária em um array de inteiros ordenado em ordem crescente. A função deve receber como parâmetros o array e o valor a ser buscado, retornando o índice do elemento caso ele seja encontrado, ou -1 caso contrário.

```python
def binary_search(arr, valor):
    start = 0
    end = len(arr) - 1
    while start <= end:
        mid = (start + end) // 2
        if arr[mid] == valor:
            return mid
        elif arr[mid] < valor:
            start = mid + 1
        else:
            end = mid - 1
    return -1
```

```python
# Exemplo de array ordenado
array = [1, 3, 5, 7, 9, 11, 13, 15]

# Testes
print(binary_search(array, 7))   # Saída esperada: 3
print(binary_search(array, 2))   # Saída esperada: -1
print(binary_search(array, 15))  # Saída esperada: 7
```

# Exercício: Busca Binária em Árvore Binária de Busca (BST)

Implemente uma função em Python que realize a busca binária em uma árvore binária de busca (BST). A função deve receber a raiz da árvore e o valor a ser buscado, retornando o nó correspondente caso ele seja encontrado, ou None caso contrário.

```python
class Node:
    def __init__(self, val):
        self.value = val
        self.left = None
        self.right = None

def binary_search(root, val):
    curr = root
    while curr:
        if val == curr.value:
            return curr
        elif val < curr.value:
            curr = curr.left
        else:
            curr = curr.right
    return None
```

```python
# Exemplo de árvore BST
root = Node(8)
root.left = Node(3)
root.right = Node(10)
root.left.esquerda = Node(1)
root.left.direita = Node(6)
root.right.direita = Node(14)
root.right.direita.left = Node(13)

# Testes
n = binary_search(root, 6)
print(n.value if n else None)  # Saída esperada: 6
n = binary_search(root, 13)
print(n.value if n else None)  # Saída esperada: 13
n = binary_search(root, 7)
print(n.value if n else None)  # Saída esperada: None
```
