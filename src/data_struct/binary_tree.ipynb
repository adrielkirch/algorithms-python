Uma **árvore binária** é uma das estruturas de dados mais fundamentais na computação. Ela organiza a informação de forma hierárquica, ao contrário de listas ou arrays que são lineares.

Aqui está um resumo completo para você dominar o conceito:

---

## 1. O que é uma Árvore Binária?

Uma árvore binária é uma estrutura de dados composta por **nós**, onde cada nó tem, no máximo, **dois filhos**, geralmente chamados de **filho da esquerda** e **filho da direita**.

Imagine uma árvore genealógica invertida: o "pai" está no topo e os "filhos" ramificam para baixo.

### Componentes Principais:

- **Raiz (Root):** O nó principal no topo da árvore.
- **Pai (Parent):** Um nó que tem sucessores.
- **Filho (Child):** Um nó que descende de outro.
- **Folha (Leaf):** Nós que não possuem filhos (as extremidades).
- **Subárvore:** Qualquer nó e seus descendentes podem ser vistos como uma árvore menor.

---

## 2. Tipos Comuns de Árvores Binárias

Nem toda árvore binária é igual. Algumas classificações importantes incluem:

- **Árvore Estrita (Full):** Cada nó tem ou 0 ou 2 filhos. Nunca apenas 1.
- **Árvore Perfeita:** Todos os níveis estão cheios e todas as folhas estão no mesmo nível.
- **Árvore Completa:** Todos os níveis estão preenchidos, exceto possivelmente o último, que deve ser preenchido da esquerda para a direita.
- **Árvore de Busca Binária (BST):** Uma árvore organizada onde o filho à esquerda é sempre **menor** que o pai, e o filho à direita é sempre **maior**.

---

## 3. Propriedades Matemáticas

Para entender a eficiência de uma árvore, usamos algumas fórmulas básicas. Se $h$ é a altura da árvore:

- **Número máximo de nós:** O total de nós em uma árvore perfeita de altura $h$ é dado por $2^{h+1} - 1$.
- **Altura mínima:** Para $n$ nós, a altura mínima é aproximadamente $\log_2(n)$. Isso é o que torna as buscas tão rápidas!

---

## 4. Formas de Percorrer a Árvore (Traversal)

Como não é uma lista simples, existem diferentes maneiras de "visitar" todos os nós:

1.  **Pré-ordem (Pre-order):** Raiz $\rightarrow$ Esquerda $\rightarrow$ Direita. (Útil para copiar a árvore).
2.  **Em-ordem (In-order):** Esquerda $\rightarrow$ Raiz $\rightarrow$ Direita. (Em uma BST, os valores saem em ordem crescente).
3.  **Pós-ordem (Post-order):** Esquerda $\rightarrow$ Direita $\rightarrow$ Raiz. (Útil para deletar a árvore).

---

## 5. Por que usar? (Vantagens)

- **Eficiência:** Em uma Árvore de Busca Binária equilibrada, operações de busca, inserção e remoção levam tempo **O(log n)**.
- **Hierarquia:** Natural para representar pastas de arquivos ou expressões matemáticas.
- **Flexibilidade:** É a base para estruturas mais complexas como Árvores AVL, Red-Black e Heaps (usadas em filas de prioridade).

---

> **Dica de Ouro:** O segredo de uma árvore binária eficiente é o **balanceamento**. Se uma árvore "pender" muito para um lado só, ela vira uma lista ligada e perde toda a sua vantagem de performance!

##  Otimização e Complexidade em Árvores
Ao implementar algoritmos de árvore, é fundamental entender a diferença entre BFS (Breadth-First Search) e DFS (Depth-First Search) em termos de complexidade de espaço e quando optar por cada um.

**BFS (Queue):**
- Utiliza uma fila para percorrer a árvore nível a nível.
- Complexidade de espaço: proporcional à largura máxima da árvore (Width).
- Melhor para árvores muito profundas (ex: listas ligadas), pois evita stack overflow.

**DFS (Recursão/Stack):**
- Utiliza recursão ou uma pilha para percorrer a árvore em profundidade.
- Complexidade de espaço: proporcional à profundidade máxima da árvore (Height/Depth).
- Melhor para árvores muito largas, pois consome menos memória.

Comente com o entrevistador que você entende essas diferenças e sabe escolher a abordagem ideal conforme a estrutura da árvore. Isso demonstra domínio avançado e atenção à otimização.

**Resumo:**
- Árvores largas: prefira DFS para menor uso de memória.
- Árvores profundas: prefira BFS para evitar stack overflow.

```python
from typing import Optional, List
from collections import deque

"""
A unidade básica é o Nó. Ele precisa guardar três coisas: o valor (dado), uma referência para o filho da esquerda e uma para o filho da direita.
"""


class Node:
    def __init__(self, value: int):
        self.value = value
        self.left: Optional["Node"] = None
        self.right: Optional["Node"] = None


def dfs_in_order(result: List[int], node: Optional[Node]):
    """
       F
     /   \
    B     G
   / \     \
  A   D     I
     / \   /
    C   E H

    ["A", "B", "C", "D", "E", "F", "G", "H", "I"]
    """
    if Node is None:
        return
    dfs_in_order(result, node.left)
    result.append(node.value)
    print(node.value)
    dfs_in_order(result, node.right)


def dfs_pre_order(result: List[int], node: Optional[Node]):
    """
       F
     /   \
    B     G
   / \     \
  A   D     I
     / \   /
    C   E H

    ["F", "B", "A", "D", "C", "E", "G", "I", "H"]
    """
    if Node is None:
        return
    result.append(node.value)
    print(node.value)
    dfs_pre_order(result, node.left)
    dfs_pre_order(result, node.right)


def dfs_post_order(result: List[int], node: Optional[Node]):
    """
       F
     /   \
    B     G
   / \     \
  A   D     I
     / \   /
    C   E H

    ["A", "C", "E", "D", "B", "H", "I", "G", "F"]
    """
    if Node is None:
        return
    dfs_post_order(result, node.left)
    dfs_post_order(result, node.right)
    result.append(node.value)
    print(node.value)

def dfs_in_order_stack(root: Optional["Node"]) -> List[int]:
    result = []
    stack = []
    node = root
    while stack or node:
        while node:
            stack.append(node)
            node = node.left
        node = stack.pop()
        result.append(node.value)
        print(node.value)
        node = node.right
    return result

def dfs_pre_order_stack(root: Optional["Node"]) -> List[int]:
    result = []
    if not root:
        return result
    stack = [root]
    while stack:
        node = stack.pop()
        result.append(node.value)
        print(node.value)
        # Primeiro direita, depois esquerda (para esquerda sair primeiro)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return result


def height(node: Optional[Node]) -> int:
    if node is None:
        return 0
    height_left = height(node.left)
    height_right = height(node.right)
    return 1 + max(height_left, height_right)


def height_queue(node: Optional[Node]) -> int:
    height = 0
    if node is None:
        return height
    queue = deque([node])
    while queue:
        levels = len(queue)
        for _ in range(levels):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        height += 1
    return height


def inverter(node: Optional[Node]) -> Optional[Node]:
    if node is None:
        return None
    node.left, node.right = inverter(node.right), inverter(node.left)
    return node


def inverter_queue(root: Optional[Node]) -> Optional[Node]:
    if root is None:
        return None
    queue = deque([root])
    while queue:
        node = queue.popleft()
        node.left, node.right = node.right, node.left
        if root.left:
            queue.append(node.left)
        if root.right:
            queue.append(node.right)
    return root

def binary_search(node: Node, target: int):
    current = node
    while current is not None:
        if target < current.value:
            current = current.left
        elif target > current.value:
            current = current.right
        else:
            return True
    return False


def binary_search_array(arr: List[int], target: int, low: int, high: int):
    if low > high:
        return None

    mid = len(arr) // 2

    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_array(arr, target, mid + 1, high)
    else:
        return binary_search_array(arr, target, low, mid + 1)


def binary_search_array_no_recursion(arr: List[int], target: int):
    left = 0
    right = len(arr) - 1

    while left < right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return None
```
